# Algorithm Analysis

## 1. Big O Notation

Big O notation expresses how an algorithm's runtime or memory usage grows relative to input size $n$. It measures worst-case performance by focusing on the dominant term as $n$ grows large.

**Simplification rules:**
- Drop constants: $O(2n)$ -> $O(n)$
- Drop coefficients: $O(5n^2)$ -> $O(n^2)$
- Drop lower-order terms: $O(n^2 + n)$ -> $O(n^2)$
- Only the dominant term matters as $n$ -> $\infty$

**Example:** $O(23n^2 + 45n + 100000)$ reduces to $O(n^2)$

## 2. Complexity Classes

| Class | Notation | Growth | Example |
|-------|----------|--------|---------|
| Constant | O(1) | Fixed | Array index access |
| Logarithmic | O(log n) | Slow | Binary search |
| Linear | O(n) | Proportional | Single loop |
| Linearithmic | O(n log n) | Moderate | Merge sort |
| Quadratic | O(n^2) | Fast | Nested loops |
| Exponential | O(2^n) | Very fast | Recursive subsets |
| Factorial | O(n!) | Explosive | Permutations |

### O(1) - Constant

Runtime does not depend on input size. The same number of operations regardless of $n$.

````python
def getFirst(nums):
    return nums[0]


def swap(nums, i, j):
    nums[i], nums[j] = nums[j], nums[i]
````

### O(log n) - Logarithmic

Input size is halved at each step. Common in divide-and-conquer search.

````python
def binarySearch(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
````

```text
n = 16: 16 -> 8 -> 4 -> 2 -> 1  (4 steps = log2(16))
n = 1024: 10 steps = log2(1024)

each step halves the search space
```

### O(n) - Linear

Runtime grows proportionally with input size. One pass through the data.

````python
def linearSearch(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1


def findMax(nums):
    res = nums[0]
    for val in nums:
        if val > res:
            res = val
    return res
````

### O(n log n) - Linearithmic

Divide-and-conquer with linear merging at each level. The lower bound for comparison-based sorting.

````python
def mergeSort(nums):
    if len(nums) <= 1:
        return nums
    mid = len(nums) // 2
    left = mergeSort(nums[:mid])
    right = mergeSort(nums[mid:])
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res
````

```text
split: log n levels of recursion
merge: O(n) work per level
total: O(n) * O(log n) = O(n log n)
```

### O(n^2) - Quadratic

Nested iteration over the input. Runtime grows with the square of input size.

````python
def bubbleSort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(n - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
    return nums


def hasDuplicateBrute(nums):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                return True
    return False
````

### O(2^n) - Exponential

Each input element doubles the number of operations. Common in recursive problems exploring all subsets.

````python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


def subsets(nums):
    if not nums:
        return [[]]
    rest = subsets(nums[1:])
    return rest + [[nums[0]] + s for s in rest]
````

```text
fibonacci call tree for n=5:

                fib(5)
              /        \
          fib(4)       fib(3)
         /     \      /     \
      fib(3)  fib(2) fib(2) fib(1)
      /   \
   fib(2) fib(1)

each level roughly doubles the number of calls
```

### O(n!) - Factorial

Generates all orderings of the input. Growth is faster than exponential.

````python
def permutations(nums):
    if len(nums) <= 1:
        return [nums[:]]
    res = []
    for i in range(len(nums)):
        nums[0], nums[i] = nums[i], nums[0]
        for perm in permutations(nums[1:]):
            res.append([nums[0]] + perm)
        nums[0], nums[i] = nums[i], nums[0]
    return res
````

```text
n=3: 3! = 6 permutations
n=5: 5! = 120 permutations
n=10: 10! = 3,628,800 permutations
```

## 3. Growth Rate Comparison

| n | O(1) | O(log n) | O(n) | O(n log n) | O(n^2) | O(2^n) | O(n!) |
|---|------|----------|------|------------|--------|--------|-------|
| 1 | 1 | 0 | 1 | 0 | 1 | 2 | 1 |
| 10 | 1 | 3 | 10 | 33 | 100 | 1,024 | 3.6M |
| 100 | 1 | 7 | 100 | 664 | 10,000 | 1.3 x 10^30 | ... |
| 1,000 | 1 | 10 | 1,000 | 9,966 | 1,000,000 | ... | ... |
| 10,000 | 1 | 13 | 10,000 | 132,877 | 100,000,000 | ... | ... |

**Ranking (best to worst):**

$$O(1) < O(\log n) < O(n) < O(n \log n) < O(n^2) < O(2^n) < O(n!)$$

## 4. Computing Time Complexity

### Step-by-Step Method

1. **Identify the input size** $n$ (array length, string length, number of nodes, etc.)
2. **Identify operations** that depend on $n$ (loops, recursive calls, built-in methods)
3. **Count the dominant operations** as a function of $n$
4. **Drop constants and lower-order terms** to get the Big O class

### Rules

| Pattern | Complexity | Reasoning |
|---------|-----------|-----------|
| Single statement | O(1) | Fixed number of operations |
| Single loop over n | O(n) | Runs n times |
| Loop with halving | O(log n) | Input cut in half each step |
| Two nested loops over n | O(n^2) | n * n iterations |
| Three nested loops | O(n^3) | n * n * n iterations |
| Sequential blocks | O(max(a, b)) | Dominant block wins |
| Loop inside loop with different sizes | O(n * m) | Depends on both inputs |
| Recursive call on n-1 | O(n) | n calls deep |
| Two recursive calls on n-1 | O(2^n) | Branching tree |
| Recursive halving | O(log n) | log n calls deep |
| Recursive halving + linear merge | O(n log n) | log n levels, O(n) per level |

## 5. Computing Space Complexity

Space complexity measures the total memory an algorithm uses relative to input size $n$. **Auxiliary space** is the extra memory beyond the input itself.

### Common Patterns

| Pattern | Space | Example |
|---------|-------|---------|
| Fixed number of variables | O(1) | Two pointers, counters |
| Array/list of size n | O(n) | Copy of input, result list |
| Hashmap/set with up to n entries | O(n) | Frequency map, seen set |
| 2D matrix n x m | O(n * m) | Grid, DP table |
| Recursion depth d | O(d) | Call stack frames |
| Recursion with branching | O(d) | Only one branch on stack at a time |

**In-place:** O(1) auxiliary space.

````python
def reverseInPlace(nums):
    left, right = 0, len(nums) - 1  # 1
    while left < right:  # n/2
        nums[left], nums[right] = nums[right], nums[left]  # 1
        left += 1  # 1
        right -= 1  # 1


# time: O(n)
# space: O(1) - only left, right
````

**Linear auxiliary space:** new data structure proportional to input.

````python
def countFrequency(s):
    freq = {}  # 1
    for ch in s:  # n
        freq[ch] = freq.get(ch, 0) + 1  # 1
    return freq  # 1


# time: O(n)
# space: O(k) where k = unique characters
````

**Recursive call stack:** each call adds a frame.

````python
def factorial(n):
    if n <= 1:  # 1
        return 1  # 1
    return n * factorial(n - 1)  # 1 + recurse


# time: O(n) - n recursive calls
# space: O(n) - n frames on the call stack
````
