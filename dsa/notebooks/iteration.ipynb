# Iteration Primitives in Python

## 1. Loop Basics

A loop is a control-flow structure that repeats a block of code while a condition holds or over a sequence of values. The two main forms are `for` (bounded, sequence-driven) and `while` (unbounded, condition-driven).

### `for` Loops

A `for` loop iterates over each element in a finite sequence. The loop variable binds to successive elements until the sequence is exhausted.

**Time complexity:** O(n) where n is the length of the sequence.
**Space complexity:** O(1) auxiliary.

````python
def printElements(nums):
    for val in nums:
        print(val)
````

### `while` Loops

A `while` loop evaluates a boolean condition before each iteration. The body executes only when the condition is true.

**Time complexity:** Depends on iteration count.
**Space complexity:** O(1) auxiliary.

````python
def countdown(n):
    while n > 0:
        print(n)
        n -= 1
````

### `for` vs `while`

| Property    | `for`                          | `while`                                  |
|-------------|--------------------------------|------------------------------------------|
| Driven by   | Sequence / iterator            | Boolean condition                        |
| Termination | Guaranteed (finite sequence)   | Not guaranteed                           |
| Indexing    | Automatic                      | Manual                                   |
| Use when    | Arrays, dicts, ranges, strings | Count not known ahead, convergence loops |

Use `for` when iterating over a known sequence. Use `while` when the number of iterations depends on a condition evaluated at each step.

## 2. Loop Control

Loop control statements alter the normal sequential flow inside a loop body. `break` exits the loop entirely, `continue` skips to the next iteration, and `pass` is a syntactic no-op placeholder.

### `break`

`break` immediately terminates the innermost enclosing loop.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{firstOccurrence} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of integers, int } target \text{ - value to find} \\
\textbf{Output: } \text{int - index of first occurrence of target, or -1} \\
\\
\textbf{function } \text{firstOccurrence}(nums, target): \\
\quad 1: \quad idx \leftarrow -1 \\
\quad 2: \quad \textbf{for } \text{each index } i \text{ in } nums: \\
\quad 3: \quad\quad \textbf{if } nums[i] \text{ equals } target: \\
\quad 4: \quad\quad\quad \text{record } idx \leftarrow i \\
\quad 5: \quad\quad\quad \textbf{break} \text{ out of the loop} \\
\quad 6: \quad \textbf{return } idx \\
\hline
\end{array}
$$

````python
def firstOccurrence(nums, target):
    idx = -1
    for i in range(len(nums)):
        if nums[i] == target:
            idx = i
            break
    return idx
````

#### Steps

1. Scan begins at index 0.
```text
nums: [4, 7, 2, 7]
       ^
       i = 0 -> 4 != 7 -> continue scanning
```

2. Target found, break.
```text
nums: [4, 7, 2, 7]
          ^
          i = 1 -> 7 == 7 -> idx = 1 -> break
```

3. Loop exited early. Return idx = 1.

### `continue`

`continue` skips the remainder of the current iteration and advances to the next.

````python
def filterEvens(nums):
    res = []
    for val in nums:
        if val % 2 != 0:
            continue
        res.append(val)
    return res
````

### `pass`

`pass` is a null statement. It does nothing and serves as a placeholder.

````python
def placeholder(nums):
    for val in nums:
        pass
````

### `for`-`else`

The `else` block on a `for` loop executes only if the loop completes without hitting a `break`.

````python
def containsTarget(nums, target):
    for val in nums:
        if val == target:
            return True
    else:
        return False
````

## 3. Loop Helpers

Loop helpers are built-in functions that generate or transform sequences to simplify common iteration patterns: 
- generating index ranges
- pairing indices with values
- traversing multiple sequences in parallel
- reversing or sorting iteration order

### `range()`

`range(start, stop, step)` produces an arithmetic sequence of integers from `start` (inclusive) to `stop` (exclusive) with a given `step`.

**Time complexity:** O(n) to iterate, where n = ceil((stop - start) / step).
**Space complexity:** O(1) (on demand).

````python
def printIndices(n):
    for i in range(n):
        print(i)
````

Common `range` variants:

```python
range(2, 7)            # 2, 3, 4, 5, 6
range(0, 10, 2)        # 0, 2, 4, 6, 8
range(5, 0, -1)        # 5, 4, 3, 2, 1
range(n - 1, -1, -1)   # n-1, n-2, ..., 1, 0 (reverse index scan)
```

### `enumerate()`

`enumerate(iterable, start=0)` pairs each element with its index.

**Time complexity:** O(n).
**Space complexity:** O(1) auxiliary.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{indexMap} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of values} \\
\textbf{Output: } \text{dict - mapping from value to its first index} \\
\\
\textbf{function } \text{indexMap}(nums): \\
\quad 1: \quad res \leftarrow \{\} \\
\quad 2: \quad \textbf{for } \text{each } (i, val) \text{ in } \text{enumerate}(nums): \\
\quad 3: \quad\quad \textbf{if } val \text{ not yet seen}: \; res[val] \leftarrow i \\
\quad 4: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def indexMap(nums):
    res = {}
    for i, val in enumerate(nums):
        if val not in res:
            res[val] = i
    return res
````

#### Steps

1. First pair yielded.
```text
nums: [a, b, a]
       ^
       (0, a) -> res = {a: 0}
```

2. New value mapped.
```text
nums: [a, b, a]
          ^
          (1, b) -> res = {a: 0, b: 1}
```

3. Duplicate skipped.
```text
nums: [a, b, a]
             ^
             (2, a) -> a already in res -> skip
```

### `zip()`

`zip(iter1, iter2, ...)` produces tuples of corresponding elements from multiple iterables, stopping at the shortest.

**Time complexity:** O(min(n, m)) for two sequences of length n and m.
**Space complexity:** O(1) auxiliary (lazy).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{vectorAdd} \\
\hline
\textbf{Input: } \text{list } a \text{ - first vector, list } b \text{ - second vector} \\
\textbf{Output: } \text{list - element-wise sum of } a \text{ and } b \\
\\
\textbf{function } \text{vectorAdd}(a, b): \\
\quad 1: \quad res \leftarrow [\,] \\
\quad 2: \quad \textbf{for } \text{each pair } (x, y) \text{ from } a, b \text{ in parallel}: \\
\quad 3: \quad\quad \text{append their sum } x + y \text{ to } res \\
\quad 4: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def vectorAdd(a, b):
    res = []
    for x, y in zip(a, b):
        res.append(x + y)
    return res
````

#### Steps

1. First pair aligned.
```text
a: [1, 2, 3]
    ^
b: [4, 5, 6]
    ^
    (1, 4) -> 1 + 4 = 5 -> res = [5]
```

2. Second pair.
```text
a: [1, 2, 3]
       ^
b: [4, 5, 6]
       ^
       (2, 5) -> 2 + 5 = 7 -> res = [5, 7]
```

3. Third pair.
```text
a: [1, 2, 3]
          ^
b: [4, 5, 6]
          ^
          (3, 6) -> 3 + 6 = 9 -> res = [5, 7, 9]
```

### `reversed()`

`reversed(seq)` returns an iterator over elements in reverse order without creating a copy.

**Time complexity:** O(n) to iterate.
**Space complexity:** O(1) (on demand).

````python
def reverseCollect(nums):
    res = []
    for val in reversed(nums):
        res.append(val)
    return res
````

#### Steps

1. Elements yielded from last to first.
```text
nums: [10, 20, 30]

reversed -> 30 -> 20 -> 10 -> exhausted
res = [30, 20, 10]
```

### `sorted()`

`sorted(iterable, key=None, reverse=False)` returns a new list of elements in sorted order. The optional `key` function extracts a comparison value from each element.

**Time complexity:** O(n log n).
**Space complexity:** O(n).

````python
def sortedTraversal(nums):
    for val in sorted(nums):
        print(val)


def sortByLength(words):
    for w in sorted(words, key=len):
        print(w)
````

#### Steps

1. Elements yielded in ascending order.
```text
nums: [3, 1, 4, 1, 5]

sorted -> [1, 1, 3, 4, 5]
val = 1 -> val = 1 -> val = 3 -> val = 4 -> val = 5
```

## 4. Common Patterns

Common patterns are recurring structural templates inside loops
- accumulating a result across iterations
- branching on conditions within the loop body
- tracking running state
- expressing filtered or transformed sequences via comprehensions.

### Accumulator Pattern

An accumulator is a variable initialized before the loop and updated each iteration to collect a running result (sum, product, count, list, etc.).

**Time complexity:** O(n).
**Space complexity:** O(1) for scalar accumulators, O(n) for list accumulators.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{prefixSums} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of numbers} \\
\textbf{Output: } \text{list - prefix sums of } nums \\
\\
\textbf{function } \text{prefixSums}(nums): \\
\quad 1: \quad res \leftarrow [\,], \; total \leftarrow 0 \\
\quad 2: \quad \textbf{for } \text{each } val \text{ in } nums: \\
\quad 3: \quad\quad \text{add value to running sum: } total \leftarrow total + val \\
\quad 4: \quad\quad \text{append } total \text{ to } res \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def prefixSums(nums):
    res = []
    total = 0
    for val in nums:
        total += val
        res.append(total)
    return res
````

#### Steps

1. First accumulation.
```text
nums: [3, 1, 4]
       ^
       total = 0 + 3 = 3 -> res = [3]
```

2. Running total grows.
```text
nums: [3, 1, 4]
          ^
          total = 3 + 1 = 4 -> res = [3, 4]
```

3. Final accumulation.
```text
nums: [3, 1, 4]
             ^
             total = 4 + 4 = 8 -> res = [3, 4, 8]
```

### Conditionals Inside Loops

Conditionals (`if`, `elif`, `else`) inside a loop body partition elements into categories or apply different logic based on element properties.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{categorizeChars} \\
\hline
\textbf{Input: } \text{str } s \text{ - input string} \\
\textbf{Output: } \text{dict - counts of digits, letters, and other characters} \\
\\
\textbf{function } \text{categorizeChars}(s): \\
\quad 1: \quad freq \leftarrow \{\text{"digits"}: 0, \; \text{"letters"}: 0, \; \text{"other"}: 0\} \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } s: \\
\quad 3: \quad\quad \textbf{if } ch \text{ is a digit: increment } freq[\text{"digits"}] \\
\quad 4: \quad\quad \textbf{else if } ch \text{ is a letter: increment } freq[\text{"letters"}] \\
\quad 5: \quad\quad \textbf{else: } \text{increment } freq[\text{"other"}] \\
\quad 6: \quad \textbf{return } freq \\
\hline
\end{array}
$$

````python
def categorizeChars(s):
    freq = {"digits": 0, "letters": 0, "other": 0}
    for ch in s:
        if ch.isdigit():
            freq["digits"] += 1
        elif ch.isalpha():
            freq["letters"] += 1
        else:
            freq["other"] += 1
    return freq
````

#### Steps

1. Letter encountered.
```text
s: "a1!"
    ^
    ch = 'a' -> isalpha -> freq = {digits: 0, letters: 1, other: 0}
```

2. Digit encountered.
```text
s: "a1!"
     ^
     ch = '1' -> isdigit -> freq = {digits: 1, letters: 1, other: 0}
```

3. Other character encountered.
```text
s: "a1!"
      ^
      ch = '!' -> else -> freq = {digits: 1, letters: 1, other: 1}
```

### Frequency Count

A frequency count builds a mapping from element to occurrence count using a dictionary accumulator.

**Time complexity:** O(n).
**Space complexity:** O(k) where k is the number of distinct elements.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{charFrequency} \\
\hline
\textbf{Input: } \text{str } s \text{ - input string} \\
\textbf{Output: } \text{dict - mapping from character to its count} \\
\\
\textbf{function } \text{charFrequency}(s): \\
\quad 1: \quad freq \leftarrow \{\} \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } s: \\
\quad 3: \quad\quad \textbf{if } ch \text{ already seen}: \; freq[ch] \leftarrow freq[ch] + 1 \\
\quad 4: \quad\quad \textbf{else: } \text{first occurrence, set } freq[ch] \leftarrow 1 \\
\quad 5: \quad \textbf{return } freq \\
\hline
\end{array}
$$

````python
def charFrequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq
````

#### Steps

1. First occurrence creates the entry.
```text
s: "abac"
    ^
    ch = 'a' -> freq = {a: 1}
```

2. New character.
```text
s: "abac"
     ^
     ch = 'b' -> freq = {a: 1, b: 1}
```

3. Repeated character increments.
```text
s: "abac"
      ^
      ch = 'a' -> freq = {a: 2, b: 1}
```

4. Final state.
```text
s: "abac"
       ^
       ch = 'c' -> freq = {a: 2, b: 1, c: 1}
```

### Running Min/Max

Track the minimum or maximum value seen so far as you scan through a sequence.

**Time complexity:** O(n).
**Space complexity:** O(1).

````python
def runningMax(nums):
    res = []
    best = float("-inf")
    for val in nums:
        best = max(best, val)
        res.append(best)
    return res
````

#### Steps

1. Track the best value seen so far at each position.
```text
nums: [2, 5, 1, 8, 3]

val = 2 -> best = 2  -> res = [2]
val = 5 -> best = 5  -> res = [2, 5]
val = 1 -> best = 5  -> res = [2, 5, 5]
val = 8 -> best = 8  -> res = [2, 5, 5, 8]
val = 3 -> best = 8  -> res = [2, 5, 5, 8, 8]
```

### Flag Pattern

A boolean flag is set before the loop and toggled inside the loop to record whether some condition was met during traversal.

**Time complexity:** O(n).
**Space complexity:** O(1).

````python
def isSorted(nums):
    flag = True
    for i in range(1, len(nums)):
        if nums[i] < nums[i - 1]:
            flag = False
            break
    return flag
````

#### Steps

1. Compare adjacent pairs until an inversion is found or the scan completes.
```text
nums: [1, 3, 2, 4]

i = 1 -> 3 >= 1 -> ok
i = 2 -> 2 < 3  -> flag = False -> break
Return False
```

### Seen Set Pattern

A set tracks previously visited elements to detect duplicates or enforce uniqueness during traversal.

**Time complexity:** O(n).
**Space complexity:** O(n).

````python
def hasDuplicate(nums):
    seen = set()
    for val in nums:
        if val in seen:
            return True
        seen.add(val)
    return False
````

#### Steps

1. Each element is checked against the set, then added.
```text
nums: [3, 1, 4, 1]

val = 3 -> seen = {3}
val = 1 -> seen = {3, 1}
val = 4 -> seen = {3, 1, 4}
val = 1 -> 1 in seen -> return True
```

### Two-Index Walk

Use two index variables to compare or process elements at different positions within the same sequence.

**Time complexity:** O(n).
**Space complexity:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{pairwiseDiff} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of numbers} \\
\textbf{Output: } \text{list - differences between each adjacent pair} \\
\\
\textbf{function } \text{pairwiseDiff}(nums): \\
\quad 1: \quad res \leftarrow [\,] \\
\quad 2: \quad \textbf{for } i \leftarrow 1 \text{ to last index } \text{len}(nums) - 1: \\
\quad 3: \quad\quad \text{compute difference: } diff \leftarrow nums[i] - nums[i - 1] \\
\quad 4: \quad\quad \text{append } diff \text{ to } res \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def pairwiseDiff(nums):
    res = []
    for i in range(1, len(nums)):
        res.append(nums[i] - nums[i - 1])
    return res
````

#### Steps

1. Adjacent elements compared via index offset.
```text
nums: [10, 30, 25, 40]

i = 1 -> 30 - 10 = 20
i = 2 -> 25 - 30 = -5
i = 3 -> 40 - 25 = 15
res = [20, -5, 15]
```

### List Comprehensions

A list comprehension builds a new list by iterating, filtering, and transforming in a single expression.

**Time complexity:** O(n).
**Space complexity:** O(k) where k is the number of elements kept.

````python
def mapTransform(nums):
    return [val * 2 for val in nums]
````

#### Steps

1. Each element is transformed in sequence.
```text
nums: [1, 2, 3]
       ^  ^  ^
       1*2=2  2*2=4  3*2=6
       result: [2, 4, 6]
```

Comprehension with filter:

````python
def filterPositive(nums):
    return [val for val in nums if val > 0]
````

```text
nums: [-1, 3, -2, 5]
        ^  ^   ^  ^
       skip keep skip keep
       result: [3, 5]
```

Comprehension with conditional expression:

````python
def clampNegatives(nums):
    return [val if val > 0 else 0 for val in nums]
````

```text
nums: [3, -1, 4, -5]
       ^   ^  ^   ^
       3   0  4   0
       result: [3, 0, 4, 0]
```

### Dictionary and Set Comprehensions

The comprehension form extends to dictionaries and sets.

````python
def invertMap(d):
    return {v: k for k, v in d.items()}


def uniqueLengths(words):
    return {len(w) for w in words}
````

#### Steps

1. Dictionary comprehension inverts key-value pairs.
```text
d: {a: 1, b: 2}

(a, 1) -> {1: a}
(b, 2) -> {1: a, 2: b}
```

2. Set comprehension collects unique lengths.
```text
words: ["hi", "bye", "ok"]

"hi"  -> len 2
"bye" -> len 3
"ok"  -> len 2 (already present)
result: {2, 3}
```

### Generator Expressions

A generator expression is a comprehension with parentheses that produces values on demand.

**Space complexity:** O(1) (no intermediate list).

````python
def sumOfSquares(nums):
    return sum(val * val for val in nums)
````

#### Steps

1. Each value is squared and fed directly into `sum` without a temporary list.
```text
nums: [1, 2, 3]

1*1 = 1 -> 2*2 = 4 -> 3*3 = 9
sum consumes on demand -> total = 14
```

## 5. Nested Iteration

Nested iteration places one loop inside another. The inner loop runs to completion for each single iteration of the outer loop. This structure is used to process pairs of elements, 2D grids, or multi-level data.

### Nested `for` Loops

Two `for` loops where the inner loop iterates fully for each step of the outer loop.

**Time complexity:** O(n * m) for outer length n and inner length m.
**Space complexity:** O(1) auxiliary (excluding output).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{cartesianPairs} \\
\hline
\textbf{Input: } \text{list } a \text{ - first list, list } b \text{ - second list} \\
\textbf{Output: } \text{list of tuples - all } (x, y) \text{ pairs with } x \text{ from } a \text{ and } y \text{ from } b \\
\\
\textbf{function } \text{cartesianPairs}(a, b): \\
\quad 1: \quad res \leftarrow [\,] \\
\quad 2: \quad \textbf{for } \text{each element } x \text{ in } a: \\
\quad 3: \quad\quad \textbf{for } \text{each element } y \text{ in } b: \\
\quad 4: \quad\quad\quad \text{append the pair } (x, y) \text{ to } res \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def cartesianPairs(a, b):
    res = []
    for x in a:
        for y in b:
            res.append((x, y))
    return res
````

#### Steps

1. Outer at first element, inner completes.
```text
a: [1, 2]    b: [a, b]
    ^
    x = 1 -> y = a -> (1, a)
              y = b -> (1, b)
```

2. Outer advances, inner completes again.
```text
a: [1, 2]    b: [a, b]
       ^
       x = 2 -> y = a -> (2, a)
                 y = b -> (2, b)
```

3. Final result.
```text
res = [(1, a), (1, b), (2, a), (2, b)]
```

### Nested `while` Loops

Two `while` loops nested, each with distinct counters and conditions. Used when iteration bounds are dynamic.

````python
def pairPrint(rows, cols):
    i = 0
    while i < rows:
        j = 0
        while j < cols:
            print(i, j)
            j += 1
        i += 1
````

#### Steps

1. Outer i = 0, inner j runs.
```text
i = 0, j = 0 -> print (0, 0)
i = 0, j = 1 -> print (0, 1)
j = 2 -> inner exits
```

2. Outer i = 1, inner j resets and runs.
```text
i = 1, j = 0 -> print (1, 0)
i = 1, j = 1 -> print (1, 1)
j = 2 -> inner exits
i = 2 -> outer exits
```

````python
def runLengthEncode(s):
    res = []
    n = len(s)
    i = 0
    while i < n:
        j = i
        while j < n and s[j] == s[i]:
            j += 1
        res.append([s[i], j - i])
        i = j
    return res
````

#### Steps

1. Outer pointer `i` marks the start of a run, inner pointer `j` scans ahead.
```text
s: "aaabbc"
    ^
    i = 0, j = 0 -> j = 1 -> j = 2 -> j = 3 (s[3] != 'a')
    res = [['a', 3]], i = 3
```

2. Next run.
```text
s: "aaabbc"
       ^
       i = 3, j = 3 -> j = 4 -> j = 5 (s[5] != 'b')
       res = [['a', 3], ['b', 2]], i = 5
```

3. Final run.
```text
s: "aaabbc"
          ^
          i = 5, j = 5 -> j = 6 (exhausted)
          res = [['a', 3], ['b', 2], ['c', 1]], i = 6 -> outer exits
```

### Nested Loops with an Accumulator

Nested loops where an accumulator collects a result from the inner product of two sequences.

**Time complexity:** O(n).
**Space complexity:** O(1) for scalar accumulator.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{dotProduct} \\
\hline
\textbf{Input: } \text{list } a \text{ - first vector, list } b \text{ - second vector} \\
\textbf{Output: } \text{number - dot product of } a \text{ and } b \\
\\
\textbf{function } \text{dotProduct}(a, b): \\
\quad 1: \quad total \leftarrow 0 \\
\quad 2: \quad \textbf{for } i \leftarrow 0 \text{ to last index } \text{len}(a) - 1: \\
\quad 3: \quad\quad \text{multiply and accumulate: } total \leftarrow total + a[i] \cdot b[i] \\
\quad 4: \quad \textbf{return } total \\
\hline
\end{array}
$$

````python
def dotProduct(a, b):
    total = 0
    for i in range(len(a)):
        total += a[i] * b[i]
    return total
````

#### Steps

1. First product accumulated.
```text
a: [2, 3, 4]    b: [5, 6, 7]
    ^                ^
    2 * 5 = 10 -> total = 10
```

2. Second product.
```text
a: [2, 3, 4]    b: [5, 6, 7]
       ^                ^
       3 * 6 = 18 -> total = 28
```

3. Third product.
```text
a: [2, 3, 4]    b: [5, 6, 7]
          ^                ^
          4 * 7 = 28 -> total = 56
```

````python
def subarraySums(nums):
    res = []
    n = len(nums)
    for i in range(n):
        total = 0
        for j in range(i, n):
            total += nums[j]
            res.append(total)
    return res
````

#### Steps

1. Outer index `i` anchors the subarray start, inner index `j` extends it.
```text
nums: [1, 2, 3]

i = 0: total = 0
       j = 0 -> total = 1 -> res = [1]
       j = 1 -> total = 3 -> res = [1, 3]
       j = 2 -> total = 6 -> res = [1, 3, 6]
```

2. Outer advances, accumulator resets.
```text
i = 1: total = 0
       j = 1 -> total = 2 -> res = [1, 3, 6, 2]
       j = 2 -> total = 5 -> res = [1, 3, 6, 2, 5]
```

3. Final subarray.
```text
i = 2: total = 0
       j = 2 -> total = 3 -> res = [1, 3, 6, 2, 5, 3]
```

### Triangle Iteration

The inner loop bound depends on the outer loop variable, producing a triangular traversal. This visits all unique pairs (i, j) where i < j.

**Time complexity:** O(n^2 / 2) = O(n^2).
**Space complexity:** O(1) auxiliary.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{uniquePairs} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of elements} \\
\textbf{Output: } \text{list of tuples - all unique pairs } (nums[i], nums[j]) \text{ where } i < j \\
\\
\textbf{function } \text{uniquePairs}(nums): \\
\quad 1: \quad res \leftarrow [\,], \; n \leftarrow \text{len}(nums) \\
\quad 2: \quad \textbf{for } i \leftarrow 0 \text{ to } n - 1: \\
\quad 3: \quad\quad \textbf{for } j \leftarrow i + 1 \text{ to } n - 1 \text{ (indices after } i \text{)}: \\
\quad 4: \quad\quad\quad \text{append } (nums[i], nums[j]) \text{ to } res \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def uniquePairs(nums):
    n = len(nums)
    res = []
    for i in range(n):
        for j in range(i + 1, n):
            res.append((nums[i], nums[j]))
    return res
````

#### Steps

1. Triangular region visited.
```text
nums: [a, b, c]

i = 0: j = 1 -> (a, b)
       j = 2 -> (a, c)
i = 1: j = 2 -> (b, c)
i = 2: (no inner iterations)

res = [(a, b), (a, c), (b, c)]
```

2. Iteration shape.
```text
    j = 0  1  2
i = 0  .  x  x
i = 1  .  .  x
i = 2  .  .  .

x = visited cell
```

### 2D / Grid Traversal

Grid traversal uses nested loops to visit every cell in a 2D matrix, typically with the outer loop over rows and the inner loop over columns.

**Time complexity:** O(rows * cols).
**Space complexity:** O(1) auxiliary.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{gridTraversal} \\
\hline
\textbf{Input: } \text{list of lists } grid \text{ - 2D matrix of values} \\
\textbf{Output: } \text{list - all elements in row-major order} \\
\\
\textbf{function } \text{gridTraversal}(grid): \\
\quad 1: \quad res \leftarrow [\,] \\
\quad 2: \quad \textbf{for } \text{each row in } grid: \\
\quad 3: \quad\quad \textbf{for } \text{each column in that row}: \\
\quad 4: \quad\quad\quad \text{append cell value } grid[row][col] \text{ to } res \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def gridTraversal(grid):
    res = []
    for row in range(len(grid)):
        for col in range(len(grid[row])):
            res.append(grid[row][col])
    return res
````

#### Steps

1. First row scanned left to right.
```text
grid:
    [1, 2, 3]
    [4, 5, 6]

row = 0, col = 0 -> 1
row = 0, col = 1 -> 2
row = 0, col = 2 -> 3
res = [1, 2, 3]
```

2. Second row scanned.
```text
row = 1, col = 0 -> 4
row = 1, col = 1 -> 5
row = 1, col = 2 -> 6
res = [1, 2, 3, 4, 5, 6]
```

### Grid Neighbor Traversal

For a given cell (row, col), iterate over its adjacent neighbors using a direction offset array. Used in flood fill, grid BFS, and connected-component labeling.

**Time complexity:** O(1) per cell.
**Space complexity:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{getNeighbors} \\
\hline
\textbf{Input: } \text{list of lists } grid, \text{ int } row, \text{ int } col \\
\textbf{Output: } \text{list - values of valid (in-bounds) neighbors} \\
\\
\textbf{function } \text{getNeighbors}(grid, row, col): \\
\quad 1: \quad dirs \leftarrow \text{offsets for up, down, left, right} \\
\quad 2: \quad res \leftarrow [\,] \\
\quad 3: \quad \textbf{for } \text{each offset } (dr, dc) \text{ in } dirs: \\
\quad 4: \quad\quad \text{compute neighbor position: } nr, nc \leftarrow row + dr, \; col + dc \\
\quad 5: \quad\quad \textbf{if } (nr, nc) \text{ is within grid bounds}: \\
\quad 6: \quad\quad\quad \text{append } grid[nr][nc] \text{ to } res \\
\quad 7: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def getNeighbors(grid, row, col):
    dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    n, m = len(grid), len(grid[0])
    res = []
    for dr, dc in dirs:
        nr, nc = row + dr, col + dc
        if 0 <= nr < n and 0 <= nc < m:
            res.append(grid[nr][nc])
    return res
````

#### Steps

1. Check all four directions from cell (1, 1) in a 3x3 grid.
```text
grid:
    [1, 2, 3]
    [4, 5, 6]
    [7, 8, 9]

center = (1, 1) -> value = 5

(-1, 0) -> (0, 1) -> value 2 (up)
( 1, 0) -> (2, 1) -> value 8 (down)
(0, -1) -> (1, 0) -> value 4 (left)
(0,  1) -> (1, 2) -> value 6 (right)

res = [2, 8, 4, 6]
```

### Column Traversal

Traverse a 2D grid column by column instead of row by row.

**Time complexity:** O(rows * cols).
**Space complexity:** O(1) auxiliary.

````python
def columnTraversal(grid):
    res = []
    if not grid:
        return res
    n, m = len(grid), len(grid[0])
    for col in range(m):
        for row in range(n):
            res.append(grid[row][col])
    return res
````

#### Steps

1. Outer loop drives columns, inner loop drives rows.
```text
grid:
    [1, 2, 3]
    [4, 5, 6]

col = 0: row = 0 -> 1, row = 1 -> 4
col = 1: row = 0 -> 2, row = 1 -> 5
col = 2: row = 0 -> 3, row = 1 -> 6

res = [1, 4, 2, 5, 3, 6]
```

### Diagonal Traversal

Traverse all diagonals of a 2D matrix. Cells on the same diagonal share the same value of (row + col) for top-left-to-bottom-right diagonals.

**Time complexity:** O(rows * cols).
**Space complexity:** O(rows * cols) for the output grouping.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{diagonalTraversal} \\
\hline
\textbf{Input: } \text{list of lists } grid \text{ - 2D matrix} \\
\textbf{Output: } \text{list of lists - elements grouped by diagonal} \\
\\
\textbf{function } \text{diagonalTraversal}(grid): \\
\quad 1: \quad diags \leftarrow \{\} \\
\quad 2: \quad \textbf{for } \text{each cell } (row, col) \text{ in } grid: \\
\quad 3: \quad\quad \text{diagonal index: } k \leftarrow row + col \\
\quad 4: \quad\quad \text{append } grid[row][col] \text{ to } diags[k] \\
\quad 5: \quad \textbf{return } diags \text{ values sorted by key} \\
\hline
\end{array}
$$

````python
def diagonalTraversal(grid):
    n, m = len(grid), len(grid[0])
    diags = {}
    for row in range(n):
        for col in range(m):
            k = row + col
            if k not in diags:
                diags[k] = []
            diags[k].append(grid[row][col])
    return [diags[k] for k in sorted(diags)]
````

#### Steps

1. Each cell is assigned to its diagonal index.
```text
grid:
    [1, 2, 3]
    [4, 5, 6]

(0,0) -> diag 0: [1]
(0,1) -> diag 1: [2]
(1,0) -> diag 1: [2, 4]
(0,2) -> diag 2: [3]
(1,1) -> diag 2: [3, 5]
(1,2) -> diag 3: [6]

result = [[1], [2, 4], [3, 5], [6]]
```

### Nested Comprehension

List comprehensions can be nested. The outer `for` corresponds to the outer loop and the inner `for` corresponds to the inner loop.

````python
def flattenGrid(grid):
    return [val for row in grid for val in row]


def cartesianComp(a, b):
    return [(x, y) for x in a for y in b]
````

#### Steps

1. Nested comprehension reads left to right: outer loop first, inner loop second.
```text
grid = [[1, 2], [3, 4]]

row = [1, 2]: val = 1 -> val = 2
row = [3, 4]: val = 3 -> val = 4

result = [1, 2, 3, 4]
```

## 6. Iterating Over Sequences

Python sequences (lists, tuples, strings) and collections (dicts, sets) are all iterable. A `for` loop over any iterable visits its elements one at a time.

### Lists

Lists are mutable, ordered sequences. Iteration visits elements in index order.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{findMax} \\
\hline
\textbf{Input: } \text{list } nums \text{ - list of numbers} \\
\textbf{Output: } \text{number - the maximum value in } nums \\
\\
\textbf{function } \text{findMax}(nums): \\
\quad 1: \quad res \leftarrow nums[0] \text{ (start with first element)} \\
\quad 2: \quad \textbf{for } \text{each } val \text{ in } nums: \\
\quad 3: \quad\quad \textbf{if } val \text{ exceeds current max}: \; res \leftarrow val \\
\quad 4: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def findMax(nums):
    res = nums[0]
    for val in nums:
        if val > res:
            res = val
    return res
````

#### Steps

1. Initialize with first element.
```text
nums: [3, 7, 2, 9]
       ^
       res = 3
```

2. Scan updates when larger value found.
```text
val = 3 -> 3 > 3? No  -> res = 3
val = 7 -> 7 > 3? Yes -> res = 7
val = 2 -> 2 > 7? No  -> res = 7
val = 9 -> 9 > 7? Yes -> res = 9
```

3. Return res = 9.

### Tuples

Tuples are immutable, ordered sequences. Iteration behaves identically to lists. Tuple unpacking in the loop header destructures each element.

````python
def extractFirst(coords):
    res = []
    for x, y in coords:
        res.append(x)
    return res
````

#### Steps

1. Tuple unpacking at each step.
```text
coords: ((1, 2), (3, 4), (5, 6))
          ^
          (x, y) = (1, 2) -> res = [1]
```

2. Continue unpacking.
```text
coords: ((1, 2), (3, 4), (5, 6))
                  ^
                  (x, y) = (3, 4) -> res = [1, 3]
```

3. Final element.
```text
coords: ((1, 2), (3, 4), (5, 6))
                          ^
                          (x, y) = (5, 6) -> res = [1, 3, 5]
```

### Strings

Strings are immutable sequences of characters. Iterating over a string yields one character per iteration.

**Time complexity:** O(n) where n is the length of the string.
**Space complexity:** O(1) auxiliary.

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{reverseString} \\
\hline
\textbf{Input: } \text{str } s \text{ - input string} \\
\textbf{Output: } \text{str - the string with characters in reverse order} \\
\\
\textbf{function } \text{reverseString}(s): \\
\quad 1: \quad res \leftarrow \text{""} \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } s: \\
\quad 3: \quad\quad \text{prepend to front: } res \leftarrow ch + res \\
\quad 4: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def reverseString(s):
    res = ""
    for ch in s:
        res = ch + res
    return res
````

#### Steps

1. Characters prepended one by one.
```text
s: "abc"

ch = 'a' -> res = "a"
ch = 'b' -> res = "ba"
ch = 'c' -> res = "cba"
```

### Dictionaries

Iterating over a dict yields its keys by default. Use `.values()` for values and `.items()` for (key, value) pairs.

**Time complexity:** O(n) where n is the number of entries.
**Space complexity:** O(1) auxiliary.

Iterating over keys:

````python
def printKeys(d):
    for k in d:
        print(k)
````

Iterating over values:

````python
def sumValues(d):
    total = 0
    for v in d.values():
        total += v
    return total
````

Iterating over key-value pairs:

````python
def invertDict(d):
    res = {}
    for k, v in d.items():
        res[v] = k
    return res
````

#### Steps

1. `.items()` yields pairs.
```text
d: {a: 1, b: 2, c: 3}

(k, v) = (a, 1) -> res = {1: a}
(k, v) = (b, 2) -> res = {1: a, 2: b}
(k, v) = (c, 3) -> res = {1: a, 2: b, 3: c}
```

### Sets

Iterating over a set yields elements in arbitrary order. Sets are unordered collections of unique values.

**Time complexity:** O(n).
**Space complexity:** O(1) auxiliary.

````python
def setTraversal(s):
    res = []
    for val in s:
        res.append(val)
    return res
````

#### Steps

1. Elements yielded in no guaranteed order.
```text
s: {3, 1, 2}

val = 1 -> res = [1]    (order may vary)
val = 2 -> res = [1, 2]
val = 3 -> res = [1, 2, 3]
```

### Reverse Index Iteration

Traverse a sequence from the last element to the first using `range(n-1, -1, -1)` or `reversed(range(n))`.

````python
def reverseIndexScan(nums):
    res = []
    for i in range(len(nums) - 1, -1, -1):
        res.append(nums[i])
    return res
````

#### Steps

1. Indices count down from last to first.
```text
nums: [a, b, c, d]

i = 3 -> d
i = 2 -> c
i = 1 -> b
i = 0 -> a
res = [d, c, b, a]
```

### Stride Iteration

Visit every k-th element using `range(start, stop, step)`.

````python
def everyKth(nums, k):
    res = []
    for i in range(0, len(nums), k):
        res.append(nums[i])
    return res
````

#### Steps

1. Step size k = 2 selects alternating elements.
```text
nums: [a, b, c, d, e], k = 2

i = 0 -> a
i = 2 -> c
i = 4 -> e
res = [a, c, e]
```

### Multi-Sequence Parallel Iteration

`zip` generalizes to more than two sequences. All sequences are walked in parallel, element by element.

````python
def weightedSum(vals, weights, scales):
    total = 0
    for v, w, s in zip(vals, weights, scales):
        total += v * w * s
    return total
````

#### Steps

1. Three sequences walked in parallel.
```text
vals:    [1, 2, 3]
weights: [4, 5, 6]
scales:  [2, 1, 3]

(1, 4, 2) -> 1*4*2 = 8   -> total = 8
(2, 5, 1) -> 2*5*1 = 10  -> total = 18
(3, 6, 3) -> 3*6*3 = 54  -> total = 72
```

### Nested Sequence Iteration

Iterate over a list of lists (or other nested iterables) by unpacking the outer structure first, then the inner.

````python
def flattenNested(nested):
    res = []
    for sublist in nested:
        for val in sublist:
            res.append(val)
    return res
````

#### Steps

1. Outer loop selects each sublist, inner loop walks its elements.
```text
nested: [[1, 2], [3], [4, 5, 6]]

sublist = [1, 2]:     val = 1 -> val = 2
sublist = [3]:        val = 3
sublist = [4, 5, 6]:  val = 4 -> val = 5 -> val = 6
res = [1, 2, 3, 4, 5, 6]
```

## 7. Reference

| Nesting Structure                  | Complexity | Use-case                                  |
|------------------------------------|------------|-------------------------------------------|
| `for`                              | O(n)       | Traverse a sequence                       |
| `while`                            | O(?)       | Repeat until condition met                |
| `for` + `break`                    | O(n)       | Early exit on match                       |
| `for` + `continue`                 | O(n)       | Skip elements by condition                |
| `for` + `else`                     | O(n)       | Detect absence after full scan            |
| `for` + conditionals               | O(n)       | Categorize or filter elements             |
| `for` in `for`                     | O(n * m)   | All pairs, Cartesian product              |
| `for` in `for` (triangular)        | O(n^2)     | Unique pairs, upper triangle              |
| `for` in `for` + conditionals      | O(n * m)   | Filtered pair generation                  |
| `while` in `while`                 | O(n * m)   | Nested traversal, dynamic bounds          |
| `while` in `while` + conditionals  | O(n * m)   | Nested traversal, dynamic bounds + filter |
| `for` in `while`                   | O(? * m)   | Dynamic outer, fixed inner                |
| `while` in `for`                   | O(n * ?)   | Fixed outer, dynamic inner                |

**Key:**
- n, m: lengths of outer and inner sequences.
- ?: iteration count depends on runtime condition.
- Complexities describe the loop structure itself, not any algorithm built from it.

### Pattern Summary

| Pattern               | State Variable        | Typical Use                              |
|-----------------------|-----------------------|------------------------------------------|
| Accumulator           | `total`, `res`        | Running sum, product, collected results  |
| Flag                  | `flag` (bool)         | Detect presence or absence of a property |
| Frequency count       | `freq` (dict)         | Count occurrences of each element        |
| Seen set              | `seen` (set)          | Detect duplicates, enforce uniqueness    |
| Running min/max       | `best`                | Track extremum seen so far               |
| Index map             | `res` (dict)          | Map values to their positions            |
| Prefix sums           | `total`, `res`        | Cumulative totals for range queries      |
| Two-index walk        | `i`, `i - 1`          | Adjacent comparison, pairwise difference |

### Comprehension Forms

| Form                    | Syntax                                       | Produces    |
|-------------------------|----------------------------------------------|-------------|
| List comprehension      | `[expr for x in iter]`                       | `list`      |
| Filtered comprehension  | `[expr for x in iter if cond]`               | `list`      |
| Conditional expression  | `[a if cond else b for x in iter]`           | `list`      |
| Dict comprehension      | `{k: v for k, v in iter}`                    | `dict`      |
| Set comprehension       | `{expr for x in iter}`                       | `set`       |
| Generator expression    | `(expr for x in iter)`                       | `generator` |
| Nested comprehension    | `[expr for x in iter1 for y in iter2]`       | `list`      |

### Loop Helpers

|----------------|----------------------------------------|-------------------------------|
