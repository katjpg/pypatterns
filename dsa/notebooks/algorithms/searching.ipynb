# Searching Algorithms

## 1. Overview

Search algorithms locate a target element within a data structure. They range from simple linear scans to divide-and-conquer strategies and windowed traversals.

| Algorithm | Time | Space | Requirement |
|-----------|------|-------|-------------|
| Linear search | O(n) | O(1) | None |
| Binary search | O(log n) | O(1) | Sorted input |
| Two pointers | O(n) | O(1) | Problem-dependent |
| Sliding window | O(n) | O(1)-O(k) | Contiguous subarray/substring |

## 2. Linear Search

- Scan elements one by one from start to end
- Return the index when the target is found
- Return -1 if the target is not in the array

| Property | Detail |
|---|---|
| **Key idea** | Sequential scan, check each element |
| **Use-case** | Unsorted data, small arrays: `[4, 2, 7, 1, 9]` |
| **Problem keywords** | "find element", "search unsorted" |
| **Time** | O(1) best, O(n) avg/worst |
| **Space** | O(1) |

````python
def linearSearch(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1
````

#### Steps

1. Scan from left, compare each element to target.
```text
nums: [4, 2, 7, 1, 9]    target = 7

i=0: 4 != 7
i=1: 2 != 7
i=2: 7 == 7 -> return 2
```

## 3. Binary Search

- Requires sorted input
- Compare target to middle element, halve the search interval each step
- Two approaches:
  1. Direct method: return immediately on match
  2. Elimination method: narrow interval to one element, then verify

| Property | Detail |
|---|---|
| **Key idea** | Halve the search interval each step |
| **Input constraint** | Array must be sorted |
| **Use-case** | Sorted arrays, boundary finding: `nums = [1, 3, 5, 7, 9, 11]` |
| **Problem keywords** | "sorted array", "find target", "search insert position", "first/last occurrence" |
| **Time** | O(log n) |
| **Space** | O(1) |

### Pseudocode (Direct Method)

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{binarySearch} \\
\hline
\textbf{function } \text{binarySearch}(nums, target): \\
\quad 1: \quad left \leftarrow 0, \; right \leftarrow n - 1 \\
\quad 2: \quad \textbf{while } left \leq right \text{ (search interval not empty)}: \\
\quad 3: \quad\quad mid \leftarrow left + (right - left) \;/\; 2 \\
\quad 4: \quad\quad \textbf{if } nums[mid] = target: \textbf{ return } mid \text{ (found)} \\
\quad 5: \quad\quad \textbf{if } nums[mid] < target: \text{ discard left half, } left \leftarrow mid + 1 \\
\quad 6: \quad\quad \textbf{else: } \text{discard right half, } right \leftarrow mid - 1 \\
\quad 7: \quad \textbf{return } {-1} \text{ (target not in array)} \\
\hline
\end{array}
$$

````python
def binarySearch(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
````

#### Steps

1. Set initial bounds.
```text
nums: [1, 3, 5, 7, 9, 11]    target = 9
       ^                 ^
     left=0           right=5
```

2. First midpoint: mid = 2, nums[2] = 5 < 9, search right half.
```text
nums: [1, 3, 5, 7, 9, 11]
                ^        ^
              left=3   right=5
```

3. Second midpoint: mid = 4, nums[4] = 9 == target -> return 4.
```text
nums: [1, 3, 5, 7, 9, 11]
                   ^
                 mid=4 -> found
```

### Key Details

**Interval:** use closed interval `[left, right]` with `left = 0`, `right = len(nums) - 1`.

**Mid calculation:**
- `mid = left + (right - left) // 2` (prevents overflow, rounds down)
- `mid = left + (right - left + 1) // 2` (rounds up, needed when `left = mid`)

**Loop condition:**
- `left <= right`: direct method, return -1 if loop exits
- `left < right`: elimination method, check `nums[left]` after loop exits

**Boundary updates:**

| Update pair | Mid formula | When to use |
|---|---|---|
| `left = mid + 1`, `right = mid` | Round down | Eliminating left side |
| `left = mid`, `right = mid - 1` | Round up | Eliminating right side |
| `left = mid + 1`, `right = mid - 1` | Either | Direct method |

### Elimination Method

Instead of returning on match, narrow the interval by ruling out the half where the target cannot be. The loop exits when a single element remains (`left == right`), then verify it.

#### Pseudocode (Elimination, Variant 1: discard left)

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{binarySearch (discard left half)} \\
\hline
\textbf{function } \text{binarySearchElim1}(nums, target): \\
\quad 1: \quad left \leftarrow 0, \; right \leftarrow n - 1 \\
\quad 2: \quad \textbf{while } left < right: \\
\quad 3: \quad\quad mid \leftarrow left + (right - left) \;/\; 2 \text{ (round down)} \\
\quad 4: \quad\quad \textbf{if } nums[mid] < target: \text{ too small, rule out } [left, mid] \\
\quad 5: \quad\quad\quad left \leftarrow mid + 1 \\
\quad 6: \quad\quad \textbf{else: } \text{target may be at } mid \text{ or left of it} \\
\quad 7: \quad\quad\quad right \leftarrow mid \\
\quad 8: \quad \text{check remaining element, return } left \text{ if match, else } {-1} \\
\hline
\end{array}
$$

#### Pseudocode (Elimination, Variant 2: discard right)

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{binarySearch (discard right half)} \\
\hline
\textbf{function } \text{binarySearchElim2}(nums, target): \\
\quad 1: \quad left \leftarrow 0, \; right \leftarrow n - 1 \\
\quad 2: \quad \textbf{while } left < right: \\
\quad 3: \quad\quad mid \leftarrow left + (right - left + 1) \;/\; 2 \text{ (round up to prevent infinite loop)} \\
\quad 4: \quad\quad \textbf{if } nums[mid] > target: \text{ too large, rule out } [mid, right] \\
\quad 5: \quad\quad\quad right \leftarrow mid - 1 \\
\quad 6: \quad\quad \textbf{else: } \text{target may be at } mid \text{ or right of it} \\
\quad 7: \quad\quad\quad left \leftarrow mid \\
\quad 8: \quad \text{check remaining element, return } left \text{ if match, else } {-1} \\
\hline
\end{array}
$$

````python
def binarySearchElim1(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        mid = left + (right - left) // 2
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid
    return left if nums[left] == target else -1
````

**Variant 2:** eliminate right side (mid rounds up to avoid infinite loop).

````python
def binarySearchElim2(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        mid = left + (right - left + 1) // 2
        if nums[mid] > target:
            right = mid - 1
        else:
            left = mid
    return left if nums[left] == target else -1
````

## 4. Two Pointers

Two pointers traverse a sequence simultaneously to solve problems in O(n) that would otherwise require O(n^2) brute force.

- **Collision pointers:** start from opposite ends, converge inward
- **Fast/slow pointers:** start from same end, move at different speeds
- **Separate pointers:** one pointer per array, move independently

### Collision Pointers

- Two pointers start at opposite ends of a sorted array
- Move inward based on a condition until they meet
- Exploits sorted order to skip invalid combinations

| Property | Detail |
|---|---|
| **Key idea** | Converge from both ends toward the center |
| **Use-case** | Pair finding in sorted arrays: `[2, 7, 11, 15]`, target = 9 |
| **Problem keywords** | "two sum sorted", "container with most water", "valid palindrome" |
| **Time** | O(n) |
| **Space** | O(1) |

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{collisionPointers} \\
\hline
\textbf{function } \text{twoSumSorted}(nums, target): \\
\quad 1: \quad \text{place pointers at opposite ends: } left \leftarrow 0, \; right \leftarrow n - 1 \\
\quad 2: \quad \textbf{while } left < right: \\
\quad 3: \quad\quad total \leftarrow nums[left] + nums[right] \\
\quad 4: \quad\quad \textbf{if } total = target: \textbf{ return } [left, right] \\
\quad 5: \quad\quad \textbf{if } total < target: \text{ need a larger sum, move } left \text{ right} \\
\quad 6: \quad\quad \textbf{else: } \text{need a smaller sum, move } right \text{ left} \\
\quad 7: \quad \textbf{return } [-1, -1] \text{ (no pair found)} \\
\hline
\end{array}
$$

````python
def twoSumSorted(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        total = nums[left] + nums[right]
        if total == target:
            return [left, right]
        elif total < target:
            left += 1
        else:
            right -= 1
    return [-1, -1]
````

#### Steps

1. Pointers at opposite ends, compute sum.
```text
nums: [2, 7, 11, 15]    target = 9
       ^           ^
     left=0     right=3
     sum = 2 + 15 = 17 > 9 -> move right inward
```

2. Adjust the larger side.
```text
nums: [2, 7, 11, 15]
       ^       ^
     left=0  right=2
     sum = 2 + 11 = 13 > 9 -> move right inward
```

3. Found.
```text
nums: [2, 7, 11, 15]
       ^   ^
     left=0 right=1
     sum = 2 + 7 = 9 == target -> return [0, 1]
```

### Fast/Slow Pointers

- Both pointers start from the same end
- Fast pointer scans ahead, slow pointer marks the write position
- Used for in-place array modification

| Property | Detail |
|---|---|
| **Key idea** | Fast scans, slow writes |
| **Use-case** | In-place dedup, remove elements: `[0, 0, 1, 1, 2]` |
| **Problem keywords** | "remove duplicates", "remove element", "move zeroes" |
| **Time** | O(n) |
| **Space** | O(1) |

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{fastSlowPointers} \\
\hline
\textbf{function } \text{removeDuplicates}(nums): \\
\quad 1: \quad slow \leftarrow 0 \\
\quad 2: \quad \textbf{for } fast \leftarrow 1 \text{ to } n - 1: \\
\quad 3: \quad\quad \textbf{if } nums[fast] \neq nums[slow]: \\
\quad 4: \quad\quad\quad \text{move } slow \text{ to next position: } slow \leftarrow slow + 1 \\
\quad 5: \quad\quad\quad \text{write new value: } nums[slow] \leftarrow nums[fast] \\
\quad 6: \quad \textbf{return } slow + 1 \text{ (length of deduplicated array)} \\
\hline
\end{array}
$$

````python
def removeDuplicates(nums):
    if len(nums) <= 1:
        return len(nums)
    slow = 0
    for fast in range(1, len(nums)):
        if nums[fast] != nums[slow]:
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1
````

#### Steps

1. Fast scans, slow marks last unique position.
```text
nums: [0, 0, 1, 1, 2]
       ^  ^
     slow fast
     0 == 0 -> skip, fast advances
```

2. New value found, slow advances and copies.
```text
nums: [0, 0, 1, 1, 2]
       ^     ^
     slow   fast
     1 != 0 -> slow=1, nums[1]=1 -> [0, 1, 1, 1, 2]
```

3. Continue.
```text
nums: [0, 1, 1, 1, 2]
          ^        ^
        slow      fast
     2 != 1 -> slow=2, nums[2]=2 -> [0, 1, 2, 1, 2]
     return slow + 1 = 3
```

### Separate Pointers

- One pointer per array, both start at index 0
- Advance pointers based on value comparison
- Arrays are typically sorted

| Property | Detail |
|---|---|
| **Key idea** | Walk two sorted arrays in parallel |
| **Use-case** | Merge, intersect sorted arrays: `[1, 3, 5]` and `[2, 3, 6]` |
| **Problem keywords** | "merge sorted arrays", "intersection", "common elements" |
| **Time** | O(n + m) |
| **Space** | O(1) auxiliary |

````python
def intersection(nums1, nums2):
    nums1.sort()
    nums2.sort()
    i, j = 0, 0
    res = []
    while i < len(nums1) and j < len(nums2):
        if nums1[i] == nums2[j]:
            if not res or nums1[i] != res[-1]:
                res.append(nums1[i])
            i += 1
            j += 1
        elif nums1[i] < nums2[j]:
            i += 1
        else:
            j += 1
    return res
````

## 5. Sliding Window

A sliding window maintains a contiguous subarray/substring using two pointers (`left`, `right`). The window expands by advancing `right` and shrinks by advancing `left`.

- **Fixed-length:** window size is constant, slide one step at a time
- **Variable-length:** window grows and shrinks to find optimal subarray

### Fixed-Length Window

- Window size `k` is given
- Slide the window one position right each step
- Update a running aggregate (sum, count, etc.) incrementally

| Property | Detail |
|---|---|
| **Key idea** | Maintain a window of size k, slide right |
| **Use-case** | Subarray averages, max in window: `nums = [2, 1, 5, 1, 3, 2]`, k = 3 |
| **Problem keywords** | "subarray of size k", "sliding window maximum", "average of subarray" |
| **Time** | O(n) |
| **Space** | O(1) |

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{fixedWindow} \\
\hline
\textbf{function } \text{maxSumSubarray}(nums, k): \\
\quad 1: \quad \text{compute initial window sum: } windowSum \leftarrow \sum nums[0..k{-}1] \\
\quad 2: \quad maxSum \leftarrow windowSum \\
\quad 3: \quad \textbf{for } right \leftarrow k \text{ to } n - 1: \\
\quad 4: \quad\quad \text{slide: add } nums[right], \text{ remove } nums[right - k] \\
\quad 5: \quad\quad maxSum \leftarrow \max(maxSum, \; windowSum) \\
\quad 6: \quad \text{return largest window sum found} \\
\hline
\end{array}
$$

````python
def maxSumSubarray(nums, k):
    windowSum = sum(nums[:k])
    maxSum = windowSum
    for right in range(k, len(nums)):
        windowSum += nums[right] - nums[right - k]
        maxSum = max(maxSum, windowSum)
    return maxSum
````

#### Steps

1. Initialize window with first k elements.
```text
nums: [2, 1, 5, 1, 3, 2]    k = 3

window: [2, 1, 5]    sum = 8
```

2. Slide: remove left element, add right element.
```text
window: [1, 5, 1]    sum = 8 - 2 + 1 = 7
window: [5, 1, 3]    sum = 7 - 1 + 3 = 9
window: [1, 3, 2]    sum = 9 - 5 + 2 = 6

maxSum = 9
```

### Variable-Length Window

- Window grows by advancing `right`
- Shrinks by advancing `left` when a constraint is violated
- Track the best valid window seen

| Property | Detail |
|---|---|
| **Key idea** | Expand right, shrink left when invalid |
| **Use-case** | Longest/shortest valid substring: `s = "abcabcbb"` |
| **Problem keywords** | "longest substring without repeating", "minimum window substring", "longest repeating character replacement" |
| **Time** | O(n) |
| **Space** | O(k) where k = unique elements in window |

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{variableWindow} \\
\hline
\textbf{function } \text{lengthOfLongestSubstring}(s): \\
\quad 1: \quad seen \leftarrow \{\}, \; left \leftarrow 0, \; maxLen \leftarrow 0 \\
\quad 2: \quad \textbf{for } right \leftarrow 0 \text{ to } n - 1: \\
\quad 3: \quad\quad ch \leftarrow s[right] \\
\quad 4: \quad\quad \textbf{if } ch \in seen \textbf{ and } seen[ch] \geq left: \\
\quad 5: \quad\quad\quad \text{shrink past duplicate: } left \leftarrow seen[ch] + 1 \\
\quad 6: \quad\quad maxLen \leftarrow \max(maxLen, \; right - left + 1) \\
\quad 7: \quad\quad \text{record position: } seen[ch] \leftarrow right \\
\quad 8: \quad \text{return longest valid window length} \\
\hline
\end{array}
$$

````python
def lengthOfLongestSubstring(s):
    seen = {}
    left = 0
    maxLen = 0
    for right in range(len(s)):
        ch = s[right]
        if ch in seen and seen[ch] >= left:
            left = seen[ch] + 1
        maxLen = max(maxLen, right - left + 1)
        seen[ch] = right
    return maxLen
````

#### Steps

1. Expand window, track last seen index of each character.
```text
s: "a b c a b c b b"
    ^
  left=0, right=0, seen={a:0}, maxLen=1
```

2. Window grows while no repeats.
```text
s: "a b c a b c b b"
    ^     ^
  left=0, right=2, seen={a:0, b:1, c:2}, maxLen=3
```

3. Repeat found, shrink from left.
```text
s: "a b c a b c b b"
          ^ ^
  s[3]='a', seen['a']=0 >= left=0 -> left=1
  seen={a:3, b:1, c:2}, maxLen=3
```

## 6. Reference

### Algorithm Comparison

| Algorithm | Best | Average | Worst | Space | Requirement |
|-----------|------|---------|-------|-------|-------------|
| Linear search | O(1) | O(n) | O(n) | O(1) | None |
| Binary search | O(1) | O(log n) | O(log n) | O(1) | Sorted |
| Two pointers | O(1) | O(n) | O(n) | O(1) | Problem-dependent |
| Sliding window | O(n) | O(n) | O(n) | O(1)-O(k) | Contiguous interval |

### Two-Pointer Patterns

| Pattern | Setup | Movement | Use-case |
|---------|-------|----------|----------|
| Collision | `left=0, right=n-1` | Converge inward | Pair sum, palindrome |
| Fast/slow | `slow=0, fast=1` | Same direction, different speed | Dedup, remove element |
| Separate | `i=0, j=0` on two arrays | Independent | Merge, intersect |

### Sliding Window Patterns

| Type | Window size | Shrink condition | Use-case |
|------|------------|-----------------|----------|
| Fixed | Constant k | Always (maintain size) | Subarray sum/avg of size k |
| Variable | Dynamic | When constraint violated | Longest/shortest valid substring |

### Binary Search Variants

| Method | Loop condition | On match | After loop | Best for |
|--------|---------------|----------|------------|----------|
| Direct | `left <= right` | Return mid | Return -1 | Simple exact search |
| Elimination 1 | `left < right` | `right = mid` | Check `nums[left]` | Left boundary |
| Elimination 2 | `left < right` | `left = mid` | Check `nums[left]` | Right boundary |
