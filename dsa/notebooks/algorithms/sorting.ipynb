# Sorting Algorithms

## 1. Overview

Sorting algorithms arrange elements of a collection in a defined order.

**By time complexity:**
- Simple O(n^2): bubble, selection, insertion
- Efficient O(n log n): merge, quick, heap
- Linear O(n): counting, bucket, radix (with constraints)

**By space:**
- In-place O(1) auxiliary: bubble, selection, insertion, shell, quick, heap
- Not in-place O(n)+: merge, counting, bucket, radix

**Stability:** a sort is stable if equal elements retain their original relative order after sorting.
- Example: two elements with the same key at positions 3 and 7 stay in that order
- Stable: bubble, insertion, merge, counting, bucket, radix
- Unstable: selection, shell, quick, heap

### Algorithm Summary

| Algorithm | Best | Average | Worst | Space | Stable |
|-----------|------|---------|-------|-------|--------|
| Bubble | O(n) | O(n^2) | O(n^2) | O(1) | Yes |
| Selection | O(n^2) | O(n^2) | O(n^2) | O(1) | No |
| Insertion | O(n) | O(n^2) | O(n^2) | O(1) | Yes |
| Shell | O(n) | O(n^2) | O(n^2) | O(1) | No |
| Merge | O(n log n) | O(n log n) | O(n log n) | O(n) | Yes |
| Quick | O(n log n) | O(n log n) | O(n^2) | O(log n) | No |
| Heap | O(n log n) | O(n log n) | O(n log n) | O(1) | No |
| Counting | O(n + k) | O(n + k) | O(n + k) | O(k) | Yes |
| Bucket | O(n) | O(n + k) | O(n^2) | O(n + k) | Yes |
| Radix | O(n * d) | O(n * d) | O(n * d) | O(n + k) | Yes |

n = number of elements, k = value range or bucket count, d = max digits

## 2. Bubble Sort

- Compare adjacent elements and swap if out of order
- After each pass, the largest unsorted element settles at the end
- Early exit if no swaps occur in a pass

| Property | Detail |
|---|---|
| **Key idea** | Repeatedly bubble the largest element to the end |
| **Use-case** | Small arrays, nearly sorted data: `[1, 3, 2, 4, 5]` |
| **Problem keywords** | "sort in place", "swap adjacent" |
| **Time** | O(n) best, O(n^2) avg/worst |
| **Space** | O(1) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{bubbleSort} \\
\hline
\textbf{function } \text{bubbleSort}(nums): \\
\quad 1: \quad \textbf{for } i \leftarrow 0 \text{ to } n - 2: \\
\quad 2: \quad\quad swapped \leftarrow \text{False} \\
\quad 3: \quad\quad \textbf{for } j \leftarrow 0 \text{ to } n - i - 2: \\
\quad 4: \quad\quad\quad \textbf{if } nums[j] > nums[j+1]: \text{ swap adjacent, mark } swapped \\
\quad 5: \quad\quad \textbf{if not } swapped: \textbf{ break} \text{ (array already sorted)} \\
\quad 6: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def bubbleSort(nums):
    n = len(nums)
    for i in range(n - 1):
        swapped = False
        for j in range(n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swapped = True
        if not swapped:
            break
    return nums
````

#### Steps

1. Compare adjacent pairs, swap if out of order.
```text
["eevee", "bulbasaur", "charmander", "abra"]
    ^          ^
  eevee > bulbasaur -> swap
-> ["bulbasaur", "eevee", "charmander", "abra"]
-> ["bulbasaur", "charmander", "abra", "eevee"]    eevee settled
```

2. Repeat on unsorted portion.
```text
["bulbasaur", "charmander", "abra", "eevee"]
-> ["bulbasaur", "abra", "charmander", "eevee"]    charmander settled
```

3. Continue until no swaps.
```text
-> ["abra", "bulbasaur", "charmander", "eevee"]
```

## 3. Selection Sort

- Divide array into sorted (left) and unsorted (right) regions
- Each pass: find the minimum in unsorted region, swap to end of sorted region
- Always scans the full unsorted region

| Property | Detail |
|---|---|
| **Key idea** | Select minimum from unsorted, place at sorted boundary |
| **Use-case** | Small arrays where swaps are expensive: `[3, 1, 2]` |
| **Problem keywords** | "find minimum", "swap to front" |
| **Time** | O(n^2) all cases |
| **Space** | O(1) |
| **Stable** | No |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{selectionSort} \\
\hline
\textbf{function } \text{selectionSort}(nums): \\
\quad 1: \quad \textbf{for } i \leftarrow 0 \text{ to } n - 2: \\
\quad 2: \quad\quad \text{find smallest value from } i \text{ to } n - 1: \; minIdx \\
\quad 3: \quad\quad \textbf{if } minIdx \neq i: \text{ swap } nums[i], nums[minIdx] \\
\quad 4: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def selectionSort(nums):
    n = len(nums)
    for i in range(n - 1):
        minIdx = i
        for j in range(i + 1, n):
            if nums[j] < nums[minIdx]:
                minIdx = j
        if minIdx != i:
            nums[i], nums[minIdx] = nums[minIdx], nums[i]
    return nums
````

#### Steps

1. Find minimum in unsorted region, swap to front.
```text
[8, 3, 5, 1]    min = 1 at index 3
^        ^
swap -> [1, 3, 5, 8]
```

2. Find minimum in remaining region.
```text
[1, 3, 5, 8]    min = 3, already in place
```

3. Array sorted.
```text
result: [1, 3, 5, 8]
```

## 4. Insertion Sort

- Maintain a sorted prefix
- For each unsorted element:
  - Shift larger sorted elements right
  - Insert into correct position
- Efficient on small or nearly-sorted arrays

| Property | Detail |
|---|---|
| **Key idea** | Shift and insert each element into sorted prefix |
| **Use-case** | Small or nearly sorted, streaming input: `[1, 2, 4, 3]` |
| **Problem keywords** | "insertion sort list", "sort nearly sorted" |
| **Time** | O(n) best, O(n^2) avg/worst |
| **Space** | O(1) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{insertionSort} \\
\hline
\textbf{function } \text{insertionSort}(nums): \\
\quad 1: \quad \textbf{for } i \leftarrow 1 \text{ to } n - 1: \\
\quad 2: \quad\quad temp \leftarrow nums[i], \; j \leftarrow i \\
\quad 3: \quad\quad \textbf{while } j > 0 \textbf{ and } nums[j-1] > temp: \\
\quad 4: \quad\quad\quad \text{shift right: } nums[j] \leftarrow nums[j-1], \; j \leftarrow j - 1 \\
\quad 5: \quad\quad \text{insert into sorted position: } nums[j] \leftarrow temp \\
\quad 6: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def insertionSort(nums):
    for i in range(1, len(nums)):
        temp = nums[i]
        j = i
        while j > 0 and nums[j - 1] > temp:
            nums[j] = nums[j - 1]
            j -= 1
        nums[j] = temp
    return nums
````

#### Steps

1. Take unsorted element, shift larger elements right, insert.
```text
["pikachu", "abra", "mewtwo"]    temp = "abra"
     ^        ^
shift "pikachu" right -> insert "abra"
-> ["abra", "pikachu", "mewtwo"]
```

2. Next element.
```text
["abra", "pikachu", "mewtwo"]    temp = "mewtwo"
              ^         ^
"mewtwo" < "pikachu" -> shift, insert
-> ["abra", "mewtwo", "pikachu"]
```

## 5. Shell Sort

- Generalization of insertion sort using a gap sequence
- Group elements by gap interval, insertion-sort each group
- Reduce gap until 1, completing a final insertion sort
- Gap sequence affects performance (common: n/2, n/4, ..., 1)

| Property | Detail |
|---|---|
| **Key idea** | Insertion sort with decreasing gap intervals |
| **Use-case** | Medium arrays, O(1) space: `[7, 2, 6, 8, 0, 4, 1, 5]` |
| **Problem keywords** | "gap sequence", "diminishing increment" |
| **Time** | O(n) best, O(n^2) worst (halving gap) |
| **Space** | O(1) |
| **Stable** | No |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{shellSort} \\
\hline
\textbf{function } \text{shellSort}(nums): \\
\quad 1: \quad gap \leftarrow n \;/\; 2 \\
\quad 2: \quad \textbf{while } gap > 0: \\
\quad 3: \quad\quad \textbf{for } i \leftarrow gap \text{ to } n - 1: \\
\quad 4: \quad\quad\quad temp \leftarrow nums[i], \; j \leftarrow i \\
\quad 5: \quad\quad\quad \textbf{while } j \geq gap \textbf{ and } nums[j - gap] > temp: \\
\quad 6: \quad\quad\quad\quad \text{shift element right by } gap: \; nums[j] \leftarrow nums[j - gap] \\
\quad 7: \quad\quad\quad \text{insert: } nums[j] \leftarrow temp \\
\quad 8: \quad\quad \text{halve the gap: } gap \leftarrow gap \;/\; 2 \\
\quad 9: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def shellSort(nums):
    n = len(nums)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = nums[i]
            j = i
            while j >= gap and nums[j - gap] > temp:
                nums[j] = nums[j - gap]
                j -= gap
            nums[j] = temp
        gap //= 2
    return nums
````

#### Steps

1. Gap = 4: group elements 4 apart, insertion-sort each group.
```text
[7, 2, 6, 8, 0, 4, 1, 5]
groups: {7,0} {2,4} {6,1} {8,5}
-> [0, 2, 1, 5, 7, 4, 6, 8]
```

2. Gap = 2: group elements 2 apart, sort each.
```text
[0, 2, 1, 5, 7, 4, 6, 8]
groups: {0,1,7,6} {2,5,4,8}
-> [0, 2, 1, 4, 6, 5, 7, 8]
```

3. Gap = 1: full insertion sort on nearly-sorted array.
```text
-> [0, 1, 2, 4, 5, 6, 7, 8]
```

## 6. Merge Sort

- Divide-and-conquer:
  - Split array in half recursively until single elements
  - Merge sorted halves back together
- Guaranteed O(n log n) regardless of input
- Requires O(n) auxiliary space for merging

| Property | Detail |
|---|---|
| **Key idea** | Recursively split, then merge sorted halves |
| **Use-case** | Stable sort, linked lists: `[node1] -> [node3] -> [node2]` |
| **Problem keywords** | "sort linked list", "merge sorted arrays", "count inversions" |
| **Time** | O(n log n) all cases |
| **Space** | O(n) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{mergeSort} \\
\hline
\textbf{function } \text{mergeSort}(nums): \\
\quad 1: \quad \textbf{if } \text{single element or empty: } \textbf{return } nums \\
\quad 2: \quad \text{split at midpoint: } mid \leftarrow n \;/\; 2 \\
\quad 3: \quad left \leftarrow \text{mergeSort}(nums[:mid]) \\
\quad 4: \quad right \leftarrow \text{mergeSort}(nums[mid:]) \\
\quad 5: \quad \textbf{return } \text{merge}(left, right) \\
\\
\textbf{function } \text{merge}(left, right): \\
\quad 1: \quad res \leftarrow [\,], \; i \leftarrow 0, \; j \leftarrow 0 \\
\quad 2: \quad \textbf{while } \text{both } i, j \text{ in bounds}: \\
\quad 3: \quad\quad \text{append smaller of } left[i], right[j] \text{ to } res \\
\quad 4: \quad \text{append remaining from } left[i:] \text{ or } right[j:] \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def mergeSort(nums):
    if len(nums) <= 1:
        return nums
    mid = len(nums) // 2
    left = mergeSort(nums[:mid])
    right = mergeSort(nums[mid:])
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res
````

#### Steps

1. Split recursively until single elements.
```text
["snorlax", "eevee", "abra", "jigglypuff"]
         /                    \
["snorlax", "eevee"]    ["abra", "jigglypuff"]
     /       \               /        \
["snorlax"] ["eevee"]   ["abra"] ["jigglypuff"]
```

2. Merge pairs in sorted order.
```text
["snorlax"] + ["eevee"] -> ["eevee", "snorlax"]
["abra"] + ["jigglypuff"] -> ["abra", "jigglypuff"]
```

3. Merge sorted halves.
```text
["eevee", "snorlax"] + ["abra", "jigglypuff"]
-> take "abra", "eevee", "jigglypuff", "snorlax"

result: ["abra", "eevee", "jigglypuff", "snorlax"]
```

## 7. Quick Sort

- Divide-and-conquer:
  - Choose a pivot element
  - Partition: elements < pivot go left, > pivot go right
  - Recurse on both sides
- Fastest in practice on average, but O(n^2) worst case with poor pivot

| Property | Detail |
|---|---|
| **Key idea** | Partition around pivot, recurse on both sides |
| **Use-case** | Large arrays in-place: `nums = [4, 7, 2, 6, 1, 3]` |
| **Problem keywords** | "sort colors", "k-th largest", "partition" |
| **Time** | O(n log n) avg, O(n^2) worst |
| **Space** | O(log n) stack |
| **Stable** | No |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{quickSort} \\
\hline
\textbf{function } \text{partition}(nums, low, high): \\
\quad 1: \quad pivot \leftarrow nums[low] \\
\quad 2: \quad i \leftarrow low, \; j \leftarrow high \\
\quad 3: \quad \textbf{while } i < j: \\
\quad 4: \quad\quad \text{move } j \text{ left to find element} < pivot \\
\quad 5: \quad\quad \text{move } i \text{ right to find element} > pivot \\
\quad 6: \quad\quad \text{swap } nums[i], nums[j] \\
\quad 7: \quad \text{place pivot in sorted position: swap } nums[i], nums[low] \\
\quad 8: \quad \textbf{return } i \\
\\
\textbf{function } \text{quickSort}(nums, low, high): \\
\quad 1: \quad \textbf{if } low < high: \\
\quad 2: \quad\quad pi \leftarrow \text{partition}(nums, low, high) \\
\quad 3: \quad\quad \text{sort left half: quickSort}(nums, low, pi - 1) \\
\quad 4: \quad\quad \text{sort right half: quickSort}(nums, pi + 1, high) \\
\hline
\end{array}
$$

````python
def partition(nums, low, high):
    pivot = nums[low]
    i, j = low, high
    while i < j:
        while i < j and nums[j] >= pivot:
            j -= 1
        while i < j and nums[i] <= pivot:
            i += 1
        nums[i], nums[j] = nums[j], nums[i]
    nums[i], nums[low] = nums[low], nums[i]
    return i


def quickSort(nums, low, high):
    if low < high:
        pi = partition(nums, low, high)
        quickSort(nums, low, pi - 1)
        quickSort(nums, pi + 1, high)
    return nums
````

#### Steps

1. Choose pivot. Scan inward from both ends.
```text
[4, 7, 2, 6, 1, 3]    pivot = 4
 ^              ^
 i              j
```

2. j finds element < pivot, i finds element > pivot. Swap.
```text
j finds 3, i finds 7
-> [4, 3, 2, 6, 1, 7]
```

3. Continue scanning.
```text
j finds 1, i finds 6
-> [4, 3, 2, 1, 6, 7]
```

4. Pointers cross. Place pivot.
```text
swap pivot -> [1, 3, 2, 4, 6, 7]
               left   ^   right
```

5. Recurse on left `[1, 3, 2]` and right `[6, 7]`.
```text
result: [1, 2, 3, 4, 6, 7]
```

## 8. Heap Sort

- Uses a max-heap to repeatedly extract the maximum element
- Two phases:
  - Build max-heap from array (siftDown from last non-leaf)
  - Extract max: swap root with last, shrink heap, siftDown
- Array storage: left child `2i+1`, right `2i+2`, parent `(i-1)//2`

| Property | Detail |
|---|---|
| **Key idea** | Build max-heap, repeatedly extract max to end |
| **Use-case** | O(n log n) with O(1) space, top-k: `heap = [9, 7, 5, 3, 1]` |
| **Problem keywords** | "k-th largest", "top k", "priority queue" |
| **Time** | O(n log n) all cases |
| **Space** | O(1) |
| **Stable** | No |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{heapSort} \\
\hline
\textbf{function } \text{siftDown}(nums, i, n): \\
\quad 1: \quad \textbf{while } \text{node } i \text{ has children}: \\
\quad 2: \quad\quad \text{find larger child between left and right} \\
\quad 3: \quad\quad \textbf{if } nums[i] < nums[larger]: \text{ swap, move down} \\
\quad 4: \quad\quad \textbf{else: break} \text{ (heap property satisfied)} \\
\\
\textbf{function } \text{heapSort}(nums): \\
\quad 1: \quad \text{build max-heap: siftDown from last non-leaf to root} \\
\quad 2: \quad \textbf{for } i \leftarrow n - 1 \textbf{ down to } 1: \\
\quad 3: \quad\quad \text{swap root (max) with } nums[i] \\
\quad 4: \quad\quad \text{siftDown}(nums, 0, i) \text{ to restore heap} \\
\quad 5: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def siftDown(nums, i, n):
    while 2 * i + 1 < n:
        left = 2 * i + 1
        right = 2 * i + 2
        larger = left
        if right < n and nums[right] > nums[left]:
            larger = right
        if nums[i] < nums[larger]:
            nums[i], nums[larger] = nums[larger], nums[i]
            i = larger
        else:
            break


def heapSort(nums):
    n = len(nums)
    for i in range((n - 2) // 2, -1, -1):
        siftDown(nums, i, n)
    for i in range(n - 1, 0, -1):
        nums[0], nums[i] = nums[i], nums[0]
        siftDown(nums, 0, i)
    return nums
````

#### Steps

1. Build max-heap by sifting down from last non-leaf.
```text
["abra", "eevee", "pikachu"]
siftDown index 0: "abra" < "pikachu" -> swap
-> ["pikachu", "eevee", "abra"]
```

2. Swap root (max) with last, siftDown to restore.
```text
swap "pikachu" with last -> ["abra", "eevee" | "pikachu"]
siftDown -> ["eevee", "abra" | "pikachu"]
```

3. Continue until heap size is 1.
```text
swap "eevee" -> ["abra" | "eevee", "pikachu"]

result: ["abra", "eevee", "pikachu"]
```

## 9. Counting Sort

- Non-comparison sort for bounded integer ranges
- Steps:
  - Count frequency of each value
  - Compute cumulative sums
  - Reverse-fill result array (preserves stability)
- Requires known, bounded value range

| Property | Detail |
|---|---|
| **Key idea** | Count occurrences, place by cumulative count |
| **Input constraint** | Bounded integer range, `k = max - min + 1` |
| **Use-case** | Small value range: `ages = [22, 18, 25, 18, 30, 22]` |
| **Problem keywords** | "sort array with known range", "frequency sort" |
| **Time** | O(n + k) |
| **Space** | O(k) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{countingSort} \\
\hline
\textbf{function } \text{countingSort}(nums): \\
\quad 1: \quad \text{find } lo, hi, \text{ compute range } k = hi - lo + 1 \\
\quad 2: \quad counts \leftarrow [0] \times k \\
\quad 3: \quad \text{tally each value: } counts[val - lo] \mathrel{+}= 1 \\
\quad 4: \quad \text{compute cumulative sums on } counts \\
\quad 5: \quad \text{reverse-fill result array using } counts \text{ for placement} \\
\quad 6: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def countingSort(nums):
    lo, hi = min(nums), max(nums)
    k = hi - lo + 1
    counts = [0] * k

    for val in nums:
        counts[val - lo] += 1

    for i in range(1, k):
        counts[i] += counts[i - 1]

    res = [0] * len(nums)
    for i in range(len(nums) - 1, -1, -1):
        val = nums[i]
        res[counts[val - lo] - 1] = val
        counts[val - lo] -= 1

    return res
````

#### Steps

1. Count frequency of each value.
```text
nums: [3, 0, 4, 2, 3, 1]    lo=0, hi=4

counts: [1, 1, 1, 2, 1]
         0  1  2  3  4
```

2. Compute cumulative sums.
```text
counts: [1, 2, 3, 5, 6]
```

3. Reverse-fill result using counts for placement.
```text
nums[5]=1 -> res[1]=1
nums[4]=3 -> res[4]=3
nums[3]=2 -> res[2]=2
...
result: [0, 1, 2, 3, 3, 4]
```

## 10. Bucket Sort

- Distribute elements into buckets by value range
- Sort each bucket individually (e.g. insertion sort)
- Concatenate sorted buckets
- Best when data is uniformly distributed

| Property | Detail |
|---|---|
| **Key idea** | Distribute into buckets, sort each, concatenate |
| **Input structure** | `buckets = [[] for _ in range(bucketCount)]` |
| **Input constraint** | Values spread evenly across a known range |
| **Use-case** | Evenly spread values: `[0.12, 0.87, 0.45, 0.63, 0.31]` |
| **Problem keywords** | "sort uniformly distributed", "bucket" |
| **Time** | O(n) best, O(n^2) worst |
| **Space** | O(n + k) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{bucketSort} \\
\hline
\textbf{function } \text{bucketSort}(nums, bucketSize): \\
\quad 1: \quad \text{find } lo, hi, \text{ compute number of buckets} \\
\quad 2: \quad \textbf{for } \text{each element: place in } buckets[(val - lo) \;/\; bucketSize] \\
\quad 3: \quad \textbf{for } \text{each bucket: sort (e.g. insertion sort)} \\
\quad 4: \quad \text{concatenate all buckets into result} \\
\quad 5: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
def bucketSort(nums, bucketSize=5):
    lo, hi = min(nums), max(nums)
    bucketCount = (hi - lo) // bucketSize + 1
    buckets = [[] for _ in range(bucketCount)]

    for val in nums:
        buckets[(val - lo) // bucketSize].append(val)

    res = []
    for bucket in buckets:
        insertionSort(bucket)
        res.extend(bucket)

    return res
````

#### Steps

1. Distribute elements into buckets by value range.
```text
nums: [39, 49, 8, 13, 22, 15, 10, 30, 5, 44]    bucketSize=10

bucket 0 [0-9]:   [8, 5]
bucket 1 [10-19]: [13, 15, 10]
bucket 2 [20-29]: [22]
bucket 3 [30-39]: [39, 30]
bucket 4 [40-49]: [49, 44]
```

2. Sort each bucket individually.
```text
bucket 0: [5, 8]
bucket 1: [10, 13, 15]
bucket 2: [22]
bucket 3: [30, 39]
bucket 4: [44, 49]
```

3. Concatenate all buckets.
```text
result: [5, 8, 10, 13, 15, 22, 30, 39, 44, 49]
```

## 11. Radix Sort

- Non-comparison sort for fixed-length integer keys
- Process digits from least significant to most significant:
  - Distribute into 10 buckets (digits 0-9)
  - Collect in bucket order
- Each pass is a stable distribution sort

| Property | Detail |
|---|---|
| **Key idea** | Sort digit by digit, least significant first |
| **Input constraint** | Non-negative integers (or fixed-length keys) |
| **Use-case** | Fixed-length keys, few digits: `[692, 503, 924, 871]` |
| **Problem keywords** | "maximum gap", "sort by digits" |
| **Time** | O(n * d) |
| **Space** | O(n + k) |
| **Stable** | Yes |

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{radixSort} \\
\hline
\textbf{function } \text{radixSort}(nums): \\
\quad 1: \quad maxDigits \leftarrow \text{digits in largest value} \\
\quad 2: \quad \textbf{for } \text{each digit position } d \text{ (ones, tens, ...):} \\
\quad 3: \quad\quad \text{place each element in } buckets[(val \;/\; 10^d) \bmod 10] \\
\quad 4: \quad\quad \text{collect from } buckets[0..9] \text{ back into } nums \\
\quad 5: \quad \textbf{return } nums \\
\hline
\end{array}
$$

````python
def radixSort(nums):
    maxDigits = len(str(max(nums)))
    for d in range(maxDigits):
        buckets = [[] for _ in range(10)]
        for val in nums:
            buckets[(val // (10**d)) % 10].append(val)
        nums = []
        for bucket in buckets:
            nums.extend(bucket)
    return nums
````

#### Steps

1. Sort by ones digit.
```text
nums: [692, 503, 924, 871, 542]
buckets: 1:[871] 2:[692,542] 3:[503] 4:[924]
-> [871, 692, 542, 503, 924]
```

2. Sort by tens digit.
```text
buckets: 0:[503] 2:[924] 4:[542] 7:[871] 9:[692]
-> [503, 924, 542, 871, 692]
```

3. Sort by hundreds digit.
```text
buckets: 5:[503,542] 6:[692] 8:[871] 9:[924]
-> [503, 542, 692, 871, 924]
```
