# Strings

## 1. String Fundamentals

A string is a finite, ordered sequence of characters. In Python, strings are immutable, subscriptable, and iterable. Any value enclosed in single or double quotes (`' '` or `" "`) is a string.

- Indexed from `0` to `len(s) - 1`
- Stored contiguously in memory (sequential access by index)
- Immutable: modifying a string creates a new object

### String Methods

#### Testing

| Method | Description |
|--------|-------------|
| `s.isalpha()` | True if all characters are alphabetic |
| `s.isdigit()` | True if all characters are digits |
| `s.isalnum()` | True if all characters are alphanumeric |
| `s.islower()` | True if all cased characters are lowercase |
| `s.isupper()` | True if all cased characters are uppercase |

#### Searching

| Method | Description |
|--------|-------------|
| `s.find(sub[, start[, end]])` | Lowest index where `sub` is found, or `-1` |
| `s.index(sub[, start[, end]])` | Like `find`, but raises `ValueError` if not found |
| `s.count(sub[, start[, end]])` | Number of non-overlapping occurrences of `sub` |
| `s.startswith(prefix)` | True if `s` starts with `prefix` |
| `s.endswith(suffix)` | True if `s` ends with `suffix` |

#### Transforming

| Method | Description |
|--------|-------------|
| `s.lower()` | Copy with all characters lowercased |
| `s.upper()` | Copy with all characters uppercased |
| `s.capitalize()` | Copy with first character capitalized |
| `s.title()` | Copy with first letter of each word capitalized |
| `s.strip([chars])` | Copy with leading and trailing characters removed |
| `s.replace(old, new[, count])` | Copy with occurrences of `old` replaced by `new` |
| `s.split([sep])` | List of substrings split by `sep` |
| `sep.join(iterable)` | Concatenate strings with `sep` between them |

## 2. String Comparison

Strings are compared lexicographically: character by character from left to right using their character encodings (ordinal values).

- If all compared characters are equal and lengths match, the strings are equal
- If a character differs, the string with the smaller character at that position is smaller
- If one string is a prefix of the other, the shorter string is smaller

**Character encoding:**
- **ASCII**: 128 characters covering English letters, digits, and symbols (`'A'` = 65, `'a'` = 97)
- **Unicode**: assigns a unique code point to every character across all writing systems
- **UTF-8**: variable-length encoding of Unicode (1 byte for ASCII, 3 bytes for CJK characters)

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{strcmp} \\
\hline
\textbf{Input: } \text{str } s1, \text{ str } s2 \\
\textbf{Output: } \text{int: } {-1} \text{ if } s1 < s2, \; 0 \text{ if equal, } 1 \text{ if } s1 > s2 \\
\\
\textbf{function } \text{strcmp}(s1, s2): \\
\quad 1: \quad \textbf{for } \text{each index } i \text{ while both strings have characters}: \\
\quad 2: \quad\quad \textbf{if } \text{ord}(s1[i]) < \text{ord}(s2[i]): \textbf{ return } {-1} \\
\quad 3: \quad\quad \textbf{if } \text{ord}(s1[i]) > \text{ord}(s2[i]): \textbf{ return } 1 \\
\quad 4: \quad \text{all characters equal; compare by length} \\
\quad 5: \quad \textbf{return } {-1} \text{ if shorter, } 1 \text{ if longer, } 0 \text{ if same} \\
\hline
\end{array}
$$

````python
def strcmp(s1, s2):
    i = 0
    while i < len(s1) and i < len(s2):
        if ord(s1[i]) < ord(s2[i]):
            return -1
        elif ord(s1[i]) > ord(s2[i]):
            return 1
        i += 1
    if len(s1) < len(s2):
        return -1
    elif len(s1) > len(s2):
        return 1
    return 0
````

#### Steps

1. Compare character encodings at each position.
```text
s1: "abc"    s2: "acc"

i = 0 -> ord('a') = 97, ord('a') = 97 -> equal, continue
i = 1 -> ord('b') = 98, ord('c') = 99 -> 98 < 99 -> return -1
```

2. If one string is a prefix of the other, the shorter string is smaller.
```text
s1: "abc"    s2: "abcde"

i = 0..2 -> all equal
len(s1) = 3 < len(s2) = 5 -> return -1
```

## 3. String Matching

String matching (pattern matching) finds all positions where a pattern string `p` occurs within a text string `T`.

- **Single-pattern matching**: find one pattern in a text (BF, Rabin-Karp, KMP, Boyer-Moore, Horspool, Sunday)
- **Multi-pattern matching**: find multiple patterns simultaneously (Trie, AC Automaton, Suffix Array)

### Single-Pattern Algorithm Comparison

| Algorithm | Preprocessing | Matching | Space | Strategy |
|-----------|--------------|----------|-------|----------|
| Brute Force | O(1) | O(n * m) | O(1) | Character-by-character |
| Rabin-Karp | O(m) | avg O(n), worst O(n * m) | O(1) | Rolling hash |
| KMP | O(m) | O(n) | O(m) | Prefix table |
| Boyer-Moore | O(m + k) | avg O(n/m), worst O(n * m) | O(k) | Bad char + good suffix |
| Horspool | O(m + k) | avg O(n), worst O(n * m) | O(k) | Bad char (simplified BM) |
| Sunday | O(m + k) | avg O(n), worst O(n * m) | O(k) | Next char after window |

n = text length, m = pattern length, k = character set size

## 4. Brute Force (BF)

The brute force algorithm aligns the pattern with each starting position in the text and compares character by character. On mismatch, the alignment shifts one position right and the pattern resets to the start.

**Time complexity:** O(n * m) worst and average.
**Space complexity:** O(1).

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{bruteForce} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{bruteForce}(T, p): \\
\quad 1: \quad i \leftarrow 0, \; j \leftarrow 0 \\
\quad 2: \quad \textbf{while } i < n \textbf{ and } j < m: \\
\quad 3: \quad\quad \textbf{if } T[i] = p[j]: \text{ advance both } i, j \\
\quad 4: \quad\quad \textbf{else: } i \leftarrow i - j + 1 \text{ (next start position), } j \leftarrow 0 \\
\quad 5: \quad \textbf{if } j = m: \textbf{ return } i - j \\
\quad 6: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def bruteForce(T, p):
    n, m = len(T), len(p)
    i, j = 0, 0
    while i < n and j < m:
        if T[i] == p[j]:
            i += 1
            j += 1
        else:
            i = i - j + 1
            j = 0
    if j == m:
        return i - j
    return -1
````

#### Steps

1. Pattern aligned at position 0, compare left to right.
```text
T: a b c a b d
p: a b d
   ^
   i=0, j=0 -> a=a -> match
   i=1, j=1 -> b=b -> match
   i=2, j=2 -> c!=d -> mismatch
```

2. Shift alignment right by one, reset pattern pointer.
```text
T: a b c a b d
p:   a b d
     ^
     i=1, j=0 -> b!=a -> mismatch
```

3. Continue shifting until match or text exhausted.
```text
T: a b c a b d
p:       a b d
         ^
         i=3, j=0 -> a=a, b=b, d=d -> full match -> return 3
```

## 5. Rabin-Karp (RK)

Rabin-Karp uses rolling hashing to compare the pattern hash against text substring hashes. Only when hashes match does it verify character by character to handle collisions.

**Rolling hash formula:**

For character set size d and prime modulus q:

- Pattern hash: $H(p) = \sum_{k=0}^{m-1} p_k \cdot d^{m-1-k} \mod q$
- Rolling update: $H(T[i{+}1..i{+}m]) = (H(T[i..i{+}m{-}1]) - T_i \cdot d^{m-1}) \cdot d + T_{i+m} \mod q$

**Time complexity:** avg O(n + m), worst O(n * m).
**Space complexity:** O(1).

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{rabinKarp} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern), int } d \text{ (base), int } q \text{ (modulus)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{rabinKarp}(T, p, d, q): \\
\quad 1: \quad \text{compute } hash\_p \text{ and first window } hash\_t \\
\quad 2: \quad power \leftarrow d^{m-1} \mod q \\
\quad 3: \quad \textbf{for } \text{each position } i \text{ in } T: \\
\quad 4: \quad\quad \textbf{if } hash\_p = hash\_t: \text{ verify character by character} \\
\quad 5: \quad\quad\quad \textbf{if } \text{confirmed match: } \textbf{return } i \\
\quad 6: \quad\quad \text{rolling update: remove } T[i] \text{, add } T[i{+}m] \\
\quad 7: \quad\quad hash\_t \leftarrow (hash\_t - \text{ord}(T[i]) \cdot power) \cdot d + \text{ord}(T[i{+}m]) \mod q \\
\quad 8: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def rabinKarp(T, p, d=256, q=101):
    n, m = len(T), len(p)
    if m == 0:
        return 0
    if n < m:
        return -1

    hash_p, hash_t = 0, 0
    for i in range(m):
        hash_p = (hash_p * d + ord(p[i])) % q
        hash_t = (hash_t * d + ord(T[i])) % q

    power = pow(d, m - 1, q)

    for i in range(n - m + 1):
        if hash_p == hash_t:
            if T[i : i + m] == p:
                return i
        if i < n - m:
            hash_t = (hash_t - power * ord(T[i])) % q
            hash_t = (hash_t * d + ord(T[i + m])) % q

    return -1
````

#### Steps

1. Compute pattern hash and first window hash.
```text
p: "cat"    d = 26    (a=0, b=1, ..., z=25)
H(cat) = 2*26^2 + 0*26^1 + 19*26^0 = 1371
```

2. Rolling update removes leftmost char, adds new rightmost char.
```text
window slides from "cat" to "ate":
H(ate) = (H(cat) - 2*26^2) * 26 + 4 = (1371 - 1352) * 26 + 4 = 498
```

3. On hash match, verify character by character to handle collisions.

## 6. KMP (Knuth-Morris-Pratt)

KMP avoids backtracking the text pointer by using a prefix table (next array). When a mismatch occurs at `p[j]`, the algorithm jumps `j` to `next[j-1]` instead of resetting to 0, because it knows the matched portion contains a prefix that equals a suffix.

**Key insight:** if `T[i..i+j] == p[0..j]` and `p[j] != T[i+j]`, and `p[0..k] == p[j-k..j]` (longest proper prefix = suffix), then `T[i+j-k..i+j] == p[0..k]` already matches. Skip to comparing `T[i+j]` with `p[k]`.

**Time complexity:** O(n + m).
**Space complexity:** O(m).

### Next Array

`next[j]` = length of the longest proper prefix of `p[0..j]` that is also a suffix.

```text
p: A B C A B C D

next[0] = 0    "A"       -> no proper prefix = suffix
next[1] = 0    "AB"      -> no match
next[2] = 0    "ABC"     -> no match
next[3] = 1    "ABCA"    -> "A" = "A"
next[4] = 2    "ABCAB"   -> "AB" = "AB"
next[5] = 3    "ABCABC"  -> "ABC" = "ABC"
next[6] = 0    "ABCABCD" -> no match
```

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{generateNext} \\
\hline
\textbf{Input: } \text{str } p \text{ (pattern of length } m \text{)} \\
\textbf{Output: } \text{array } nxt \text{: longest proper prefix-suffix length for each } p[0..j] \\
\\
\textbf{function } \text{generateNext}(p): \\
\quad 1: \quad nxt \leftarrow \text{array of } m \text{ zeros} \\
\quad 2: \quad left \leftarrow 0 \\
\quad 3: \quad \textbf{for } right \leftarrow 1 \text{ to } m - 1: \\
\quad 4: \quad\quad \textbf{while } left > 0 \textbf{ and } p[left] \neq p[right]: \\
\quad 5: \quad\quad\quad \text{fall back: } left \leftarrow nxt[left - 1] \\
\quad 6: \quad\quad \textbf{if } p[left] = p[right]: \; left \leftarrow left + 1 \\
\quad 7: \quad\quad nxt[right] \leftarrow left \\
\quad 8: \quad \textbf{return } nxt \\
\hline
\end{array}
$$

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{kmp} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{kmp}(T, p): \\
\quad 1: \quad nxt \leftarrow \text{generateNext}(p) \\
\quad 2: \quad j \leftarrow 0 \\
\quad 3: \quad \textbf{for } \text{each index } i \text{ in } T \text{ (never backtracks)}: \\
\quad 4: \quad\quad \textbf{while } j > 0 \textbf{ and } T[i] \neq p[j]: \; j \leftarrow nxt[j - 1] \\
\quad 5: \quad\quad \textbf{if } T[i] = p[j]: \; j \leftarrow j + 1 \\
\quad 6: \quad\quad \textbf{if } j = m: \textbf{ return } i - m + 1 \\
\quad 7: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def generateNext(p):
    m = len(p)
    nxt = [0] * m
    left = 0
    for right in range(1, m):
        while left > 0 and p[left] != p[right]:
            left = nxt[left - 1]
        if p[left] == p[right]:
            left += 1
        nxt[right] = left
    return nxt


def kmp(T, p):
    n, m = len(T), len(p)
    if m == 0:
        return 0
    nxt = generateNext(p)
    j = 0
    for i in range(n):
        while j > 0 and T[i] != p[j]:
            j = nxt[j - 1]
        if T[i] == p[j]:
            j += 1
        if j == m:
            return i - m + 1
    return -1
````

#### Steps

1. Build next array for pattern `"ABCABD"`.
```text
p:    A  B  C  A  B  D
next: 0  0  0  1  2  0
```

2. Match against text. On mismatch at j=5, use next[4]=2 to skip.
```text
T: A B C A B C A B D
p: A B C A B D
               ^
   j=5 mismatch -> j = next[4] = 2
   now compare T[5] with p[2] -> C=C -> continue

T: A B C A B C A B D
p:       A B C A B D
                     ^
   j=5 -> D=D -> full match at index 3
```

3. Text pointer `i` never backtracks. Only `j` adjusts via the next array.

## 7. Boyer-Moore (BM)

Boyer-Moore compares the pattern from right to left and uses two heuristic rules to skip positions:

**Bad character rule:** when a mismatch occurs at `p[j]`, find the rightmost occurrence of `T[i+j]` in `p`. Shift the pattern to align that occurrence with the mismatched position.

**Good suffix rule:** when a mismatch occurs after matching a suffix of `p`, find the rightmost copy of that suffix (or a prefix matching its tail) in `p`, and shift to align it.

The pattern shifts by the maximum of both rules.

**Time complexity:** avg O(n/m), worst O(n * m).
**Space complexity:** O(m + k) where k = character set size.

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{boyerMoore} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{boyerMoore}(T, p): \\
\quad 1: \quad \text{build bad character table } bc \text{ and good suffix table } gs \\
\quad 2: \quad i \leftarrow 0 \\
\quad 3: \quad \textbf{while } i \leq n - m: \\
\quad 4: \quad\quad j \leftarrow m - 1 \text{ (compare right to left)} \\
\quad 5: \quad\quad \textbf{while } j \geq 0 \textbf{ and } T[i{+}j] = p[j]: \; j \leftarrow j - 1 \\
\quad 6: \quad\quad \textbf{if } j < 0: \textbf{ return } i \\
\quad 7: \quad\quad bad\_move \leftarrow j - bc[T[i{+}j]] \\
\quad 8: \quad\quad good\_move \leftarrow gs[j] \\
\quad 9: \quad\quad i \leftarrow i + \max(bad\_move, \; good\_move) \\
\quad 10: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def generateBadCharTable(p):
    bc = {}
    for i in range(len(p)):
        bc[p[i]] = i
    return bc


def generateSuffix(p):
    m = len(p)
    suffix = [0] * m
    suffix[m - 1] = m
    for i in range(m - 2, -1, -1):
        j = i
        while j >= 0 and p[j] == p[m - 1 - (i - j)]:
            j -= 1
        suffix[i] = i - j
    return suffix


def generateGoodSuffix(p):
    m = len(p)
    gs = [m] * m
    suffix = generateSuffix(p)
    j = 0
    for i in range(m - 1, -1, -1):
        if suffix[i] == i + 1:
            while j < m - 1 - i:
                if gs[j] == m:
                    gs[j] = m - 1 - i
                j += 1
    for i in range(m - 1):
        gs[m - 1 - suffix[i]] = m - 1 - i
    return gs


def boyerMoore(T, p):
    n, m = len(T), len(p)
    if m == 0 or n < m:
        return -1
    bc = generateBadCharTable(p)
    gs = generateGoodSuffix(p)
    i = 0
    while i <= n - m:
        j = m - 1
        while j >= 0 and T[i + j] == p[j]:
            j -= 1
        if j < 0:
            return i
        bad_move = j - bc.get(T[i + j], -1)
        good_move = gs[j]
        i += max(bad_move, good_move)
    return -1
````

#### Steps

1. Compare from right to left. Bad character not in pattern -> large shift.
```text
T: H E R E _ I S _ A _ S I M P L E _ E X A M P L E
p: E X A M P L E
               ^
   j=6: T[6]='S' != p[6]='E'
   'S' not in pattern -> shift by j+1 = 7
```

2. Bad character found in pattern -> align its rightmost occurrence.
```text
T: ... _ E X A M P L E
p:       E X A M P L E
                       ^
   full match -> return position
```

## 8. Horspool

Horspool is a simplified Boyer-Moore that uses only the bad character rule, keyed on the rightmost character of the current text window (`T[i + m - 1]`).

- If `T[i + m - 1]` appears in `p[0..m-2]`: shift to align its rightmost occurrence
- If not in pattern: shift by `m`

**Time complexity:** avg O(n), worst O(n * m).
**Space complexity:** O(m + k).

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{horspool} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{horspool}(T, p): \\
\quad 1: \quad \text{build shift table from } p[0..m{-}2] \\
\quad 2: \quad i \leftarrow 0 \\
\quad 3: \quad \textbf{while } i \leq n - m: \\
\quad 4: \quad\quad j \leftarrow m - 1 \text{ (compare right to left)} \\
\quad 5: \quad\quad \textbf{while } j \geq 0 \textbf{ and } T[i{+}j] = p[j]: \; j \leftarrow j - 1 \\
\quad 6: \quad\quad \textbf{if } j < 0: \textbf{ return } i \\
\quad 7: \quad\quad \text{shift by window's last char: } i \leftarrow i + table[T[i{+}m{-}1]] \\
\quad 8: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def horspoolShiftTable(p):
    m = len(p)
    table = {}
    for i in range(m - 1):
        table[p[i]] = m - 1 - i
    return table


def horspool(T, p):
    n, m = len(T), len(p)
    if m == 0 or n < m:
        return -1
    table = horspoolShiftTable(p)
    i = 0
    while i <= n - m:
        j = m - 1
        while j >= 0 and T[i + j] == p[j]:
            j -= 1
        if j < 0:
            return i
        i += table.get(T[i + m - 1], m)
    return -1
````

#### Steps

1. Compare right to left. On mismatch, look up last window character.
```text
T: a b c a b d a b c
p: a b d
       ^
   j=2: mismatch. Window last char = T[2] = 'c'
   'c' not in shift table -> shift by m = 3
```

2. Next window.
```text
T: a b c a b d a b c
p:       a b d
             ^
   j=2: d=d, j=1: b=b, j=0: a=a -> match at index 3
```

## 9. Sunday

Sunday matches left to right. On mismatch, it examines the character just after the current window (`T[i + m]`) to determine the shift distance.

- If `T[i + m]` appears in pattern: shift to align its rightmost occurrence
- If not in pattern: shift by `m + 1`

**Time complexity:** avg O(n), worst O(n * m).
**Space complexity:** O(m + k).

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{sunday} \\
\hline
\textbf{Input: } \text{str } T \text{ (text), str } p \text{ (pattern)} \\
\textbf{Output: } \text{int: starting index of first match, or } {-1} \\
\\
\textbf{function } \text{sunday}(T, p): \\
\quad 1: \quad \text{build shift table from all of } p \\
\quad 2: \quad i \leftarrow 0 \\
\quad 3: \quad \textbf{while } i \leq n - m: \\
\quad 4: \quad\quad j \leftarrow 0 \text{ (compare left to right)} \\
\quad 5: \quad\quad \textbf{while } j < m \textbf{ and } T[i{+}j] = p[j]: \; j \leftarrow j + 1 \\
\quad 6: \quad\quad \textbf{if } j = m: \textbf{ return } i \\
\quad 7: \quad\quad \text{shift by next char past window: } i \leftarrow i + table[T[i{+}m]] \\
\quad 8: \quad \textbf{return } {-1} \\
\hline
\end{array}
$$

````python
def sundayShiftTable(p):
    m = len(p)
    table = {}
    for i in range(m):
        table[p[i]] = m - i
    return table


def sunday(T, p):
    n, m = len(T), len(p)
    if m == 0:
        return 0
    if n < m:
        return -1
    table = sundayShiftTable(p)
    i = 0
    while i <= n - m:
        j = 0
        while j < m and T[i + j] == p[j]:
            j += 1
        if j == m:
            return i
        if i + m >= n:
            return -1
        i += table.get(T[i + m], m + 1)
    return -1
````

#### Steps

1. Compare left to right. On mismatch, check next char after window.
```text
T: a b c a b d
p: a b d
   ^
   j=0: a=a, j=1: b=b, j=2: c!=d -> mismatch
   next char = T[3] = 'a'
   'a' in table -> shift by table['a'] = 3
```

2. New window at index 3.
```text
T: a b c a b d
p:       a b d
         ^
   j=0: a=a, j=1: b=b, j=2: d=d -> match at index 3
```

## 10. Trie (Prefix Tree)

A trie is a tree structure for storing and searching collections of strings. Each node represents a character, and the path from root to node forms a string. Strings sharing a common prefix share the same path.

**Properties:**
- Root node stores no character
- Each node stores one character and links to child nodes
- All children of a node have distinct characters
- An end-of-word marker distinguishes complete words from prefixes

**Operations:**
- Insert: O(n), Search: O(n), Prefix check: O(n), where n = word length

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{Trie insert} \\
\hline
\textbf{function } \text{insert}(word): \\
\quad 1: \quad cur \leftarrow root \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } word: \\
\quad 3: \quad\quad \textbf{if } ch \notin cur.\text{children}: \text{ create new node} \\
\quad 4: \quad\quad cur \leftarrow cur.\text{children}[ch] \\
\quad 5: \quad cur.\text{isEnd} \leftarrow \text{True} \\
\hline
\end{array}
$$

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{Trie search} \\
\hline
\textbf{function } \text{search}(word): \\
\quad 1: \quad cur \leftarrow root \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } word: \\
\quad 3: \quad\quad \textbf{if } ch \notin cur.\text{children}: \textbf{ return } \text{False} \\
\quad 4: \quad\quad cur \leftarrow cur.\text{children}[ch] \\
\quad 5: \quad \textbf{return } cur.\text{isEnd} \\
\hline
\end{array}
$$

````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isEnd = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        cur = self.root
        for ch in word:
            if ch not in cur.children:
                cur.children[ch] = TrieNode()
            cur = cur.children[ch]
        cur.isEnd = True

    def search(self, word):
        cur = self.root
        for ch in word:
            if ch not in cur.children:
                return False
            cur = cur.children[ch]
        return cur.isEnd

    def startsWith(self, prefix):
        cur = self.root
        for ch in prefix:
            if ch not in cur.children:
                return False
            cur = cur.children[ch]
        return True
````

#### Steps

1. Insert `"abc"`, `"abd"`, `"b"` into an empty trie.
```text
insert "abc":   root -> a -> b -> c (end)
insert "abd":   root -> a -> b -> c (end)
                              \-> d (end)
insert "b":     root -> a -> b -> ...
                 \-> b (end)
```

2. Search `"abc"` -> follow root -> a -> b -> c, isEnd = True -> found.
3. Search `"ab"` -> follow root -> a -> b, isEnd = False -> not found.
4. startsWith `"ab"` -> follow root -> a -> b, exists -> True.

### Applications

- Autocomplete and search suggestions
- Prefix matching and prefix counting
- Spell checking
- Longest common prefix
- Lexicographic sorting of strings

## 11. AC Automaton (Aho-Corasick)

The AC automaton finds all occurrences of multiple pattern strings in a single scan of the text. It combines a trie with KMP-style fail pointers.

**Construction:**
1. Build a trie from all pattern strings
2. Construct fail pointers via BFS: each node's fail pointer points to the longest proper suffix of its path that exists as a prefix in the trie

**Matching:**
- Scan the text character by character
- On match, follow child pointers; on mismatch, follow fail pointers
- At each node, walk the fail chain to check for completed patterns

**Time complexity:** O(n + m + k) where n = text length, m = total pattern length, k = number of matches.
**Space complexity:** O(m).

### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{AC Automaton buildFailPointers} \\
\hline
\textbf{function } \text{buildFailPointers}(): \\
\quad 1: \quad \text{set fail of all root children to root, enqueue them} \\
\quad 2: \quad \textbf{while } \text{queue not empty}: \\
\quad 3: \quad\quad cur \leftarrow \text{dequeue} \\
\quad 4: \quad\quad \textbf{for } \text{each } (ch, child) \text{ in } cur.\text{children}: \\
\quad 5: \quad\quad\quad \text{walk } cur\text{.fail chain for a node with } ch \text{ child} \\
\quad 6: \quad\quad\quad child.\text{fail} \leftarrow \text{that child, or root if none found} \\
\quad 7: \quad\quad\quad \text{enqueue } child \\
\hline
\end{array}
$$

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{AC Automaton search} \\
\hline
\textbf{function } \text{search}(text): \\
\quad 1: \quad node \leftarrow root, \; res \leftarrow [\,] \\
\quad 2: \quad \textbf{for } \text{each character } ch \text{ in } text: \\
\quad 3: \quad\quad \textbf{while } node \neq root \textbf{ and } ch \notin node.\text{children}: \; node \leftarrow node.\text{fail} \\
\quad 4: \quad\quad \textbf{if } ch \in node.\text{children}: \; node \leftarrow node.\text{children}[ch] \\
\quad 5: \quad\quad \text{walk fail chain from } node\text{, collect completed words} \\
\quad 6: \quad \textbf{return } res \\
\hline
\end{array}
$$

````python
class ACNode:
    def __init__(self):
        self.children = {}
        self.fail = None
        self.isEnd = False
        self.word = ""


class ACAuto:
    def __init__(self):
        self.root = ACNode()

    def addWord(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = ACNode()
            node = node.children[ch]
        node.isEnd = True
        node.word = word

    def buildFailPointers(self):
        queue = []
        for child in self.root.children.values():
            child.fail = self.root
            queue.append(child)
        while queue:
            cur = queue.pop(0)
            for ch, child in cur.children.items():
                fail = cur.fail
                while fail and ch not in fail.children:
                    fail = fail.fail
                child.fail = (
                    fail.children[ch] if fail and ch in fail.children else self.root
                )
                queue.append(child)

    def search(self, text):
        res = []
        node = self.root
        for ch in text:
            while node is not self.root and ch not in node.children:
                node = node.fail
            if ch in node.children:
                node = node.children[ch]
            temp = node
            while temp is not self.root:
                if temp.isEnd:
                    res.append(temp.word)
                temp = temp.fail
        return res
````

#### Steps

1. Build trie from patterns `["she", "he", "her"]`.
```text
    root
   /    \
  s      h
  |      |
  h      e (end: "he")
  |       \
  e (end)  r (end: "her")
  (end: "she")
```

2. Construct fail pointers.
```text
s -> root    (root's child)
h -> root    (root's child)
sh -> h      (root has 'h' child)
he -> root   (no 'e' child on root's fail chain)
she -> he    ('h' node has 'e' child)
her -> root  (no 'r' on fail chain)
```

3. Search text `"yasherhs"`.
```text
y -> root (no match)
a -> root (no match)
s -> s node
h -> sh node
e -> she node -> found "she", fail chain -> he node -> found "he"
r -> her node -> found "her"
h -> h node (fail from her -> root -> h)
s -> s node
result: ["she", "he", "her"]
```

## 12. Suffix Array

A suffix array stores all suffixes of a string sorted in lexicographic order. `SA[i]` gives the starting index of the i-th smallest suffix.

**Construction methods:**
- Naive: generate and sort all suffixes, O(n^2 log n)
- Doubling: radix sort with increasing key lengths, O(n log n)
- DC3/Skew: linear time O(n)

**LCP array:** `LCP[i]` = length of the longest common prefix between `SA[i]` and `SA[i-1]`. Used for finding longest repeated substrings and counting distinct substrings.

### Example

```text
S = "banana"

Suffixes sorted lexicographically:
  SA[0] = 5  ->  "a"
  SA[1] = 3  ->  "ana"
  SA[2] = 1  ->  "anana"
  SA[3] = 0  ->  "banana"
  SA[4] = 4  ->  "na"
  SA[5] = 2  ->  "nana"

SA = [5, 3, 1, 0, 4, 2]

LCP array:
  LCP[0] = 0   (no predecessor)
  LCP[1] = 1   "a" vs "ana" -> "a"
  LCP[2] = 3   "ana" vs "anana" -> "ana"
  LCP[3] = 0   "anana" vs "banana" -> ""
  LCP[4] = 0   "banana" vs "na" -> ""
  LCP[5] = 2   "na" vs "nana" -> "na"
```

### Pseudocode (Doubling Algorithm)

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{buildSuffixArray} \\
\hline
\textbf{Input: } \text{str } s \text{ of length } n \\
\textbf{Output: } \text{array } SA \text{ of suffix indices in lexicographic order} \\
\\
\textbf{function } \text{buildSuffixArray}(s): \\
\quad 1: \quad rank[i] \leftarrow \text{ord}(s[i]) \text{ for each } i \\
\quad 2: \quad k \leftarrow 1 \\
\quad 3: \quad \textbf{repeat}: \\
\quad 4: \quad\quad \text{sort by two-key pair } (rank[i], \; rank[i{+}k]) \\
\quad 5: \quad\quad \text{recompute ranks from sorted order} \\
\quad 6: \quad\quad \textbf{if } \text{all ranks unique: } \textbf{break} \\
\quad 7: \quad\quad k \leftarrow k \times 2 \\
\quad 8: \quad \textbf{return } SA \\
\hline
\end{array}
$$

````python
def buildSuffixArray(s):
    n = len(s)
    k = 1
    rank = [ord(c) for c in s]
    sa = list(range(n))

    def sortKey(x):
        first = rank[x]
        second = rank[x + k] if x + k < n else -1
        return (first, second)

    def isDifferent(i):
        if rank[sa[i]] != rank[sa[i - 1]]:
            return True
        cur = rank[sa[i] + k] if sa[i] + k < n else -1
        prev = rank[sa[i - 1] + k] if sa[i - 1] + k < n else -1
        return cur != prev

    while True:
        sa.sort(key=sortKey)
        tmp = [0] * n
        for i in range(1, n):
            tmp[sa[i]] = tmp[sa[i - 1]] + isDifferent(i)
        rank = tmp[:]
        if rank[sa[-1]] == n - 1:
            break
        k <<= 1
    return sa
````
