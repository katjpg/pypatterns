# Linked Lists

## 1. Linked List Fundamentals

A linked list is a linear data structure where elements are stored in nodes connected by pointers. Each node holds a value and a reference to the next node. Unlike arrays, linked list nodes need not be contiguous in memory.

- Sequential access only (no random access by index)
- Dynamic size: nodes allocated on demand
- Insertion and deletion at a known position are O(1) pointer operations
- Extra space per node for the pointer field

### Node and List Structure

A singly linked list node stores a value and a pointer to the next node. The list itself tracks only the head pointer.

````python
class ListNode:
    def __init__(self, val=0, nxt=None):
        self.val = val
        self.next = nxt


class LinkedList:
    def __init__(self):
        self.head = None
````

### Linked List Types

| Type | Structure | Use-case |
|------|-----------|----------|
| Singly linked | Each node points to next | Basic list operations |
| Doubly linked | Each node points to prev and next | Bidirectional traversal |
| Circular | Tail points back to head | Round-robin, cyclic access |

## 2. Basic Operations

### Create

Build a linked list from a sequence by creating nodes and chaining them together.

**Time:** O(n). **Space:** O(n).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{create} \\
\hline
\textbf{Input: } \text{sequence } data \\
\textbf{Output: } \text{linked list built from } data \\
\\
\textbf{function } \text{create}(data): \\
\quad 1: \quad head \leftarrow \text{new node from } data[0] \\
\quad 2: \quad cur \leftarrow head \\
\quad 3: \quad \textbf{for } \text{each remaining element in } data: \\
\quad 4: \quad\quad cur.next \leftarrow \text{new node from element} \\
\quad 5: \quad\quad \text{advance } cur \leftarrow cur.next \\
\hline
\end{array}
$$

````python
def create(self, data):
    if not data:
        return
    self.head = ListNode(data[0])
    cur = self.head
    for i in range(1, len(data)):
        cur.next = ListNode(data[i])
        cur = cur.next
````

### Length

Traverse from head to end, counting nodes.

**Time:** O(n). **Space:** O(1).

````python
def length(self):
    cnt = 0
    cur = self.head
    while cur:
        cnt += 1
        cur = cur.next
    return cnt
````

### Search

Linear scan from head, comparing each node's value to the target.

**Time:** O(n). **Space:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{find} \\
\hline
\textbf{Input: } \text{value } val \\
\textbf{Output: } \text{node with value } val \text{, or None} \\
\\
\textbf{function } \text{find}(val): \\
\quad 1: \quad cur \leftarrow head \\
\quad 2: \quad \textbf{while } cur \neq \text{None}: \\
\quad 3: \quad\quad \textbf{if } cur.val = val: \textbf{ return } cur \\
\quad 4: \quad\quad \text{traverse: } cur \leftarrow cur.next \\
\quad 5: \quad \textbf{return } \text{None} \\
\hline
\end{array}
$$

````python
def find(self, val):
    cur = self.head
    while cur:
        if cur.val == val:
            return cur
        cur = cur.next
    return None
````

### Insert at Index

Traverse to the predecessor of the target position, then splice in the new node.

**Time:** O(n). **Space:** O(1).

#### Steps

1. To insert at index 2 in list `1 -> 3 -> 5`:
```text
before:  1 -> 3 -> 5 -> None
              ^
              predecessor (index 1)

new node: 9

after:   1 -> 3 -> 9 -> 5 -> None
              |         ^
              +-> next = new node
                  new.next = old next
```

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{insertAt} \\
\hline
\textbf{Input: } \text{int } index \text{, value } val \\
\textbf{Output: } \text{new node inserted at position } index \\
\\
\textbf{function } \text{insertAt}(index, val): \\
\quad 1: \quad \textbf{if } index \leq 0: \text{ insert before head, } head \leftarrow \text{new node} \\
\quad 2: \quad \text{traverse to predecessor: } cur \leftarrow \text{node at position } index - 1 \\
\quad 3: \quad node \leftarrow \text{new node with value } val \\
\quad 4: \quad node.next \leftarrow cur.next \\
\quad 5: \quad cur.next \leftarrow node \\
\hline
\end{array}
$$

````python
def insertAt(self, index, val):
    if index <= 0:
        node = ListNode(val)
        node.next = self.head
        self.head = node
        return
    cnt = 0
    cur = self.head
    while cur and cnt < index - 1:
        cur = cur.next
        cnt += 1
    if not cur:
        return
    node = ListNode(val)
    node.next = cur.next
    cur.next = node
````

### Update at Index

Traverse to the target position and overwrite the value.

**Time:** O(n). **Space:** O(1).

````python
def change(self, index, val):
    cnt = 0
    cur = self.head
    while cur and cnt < index:
        cur = cur.next
        cnt += 1
    if not cur:
        return
    cur.val = val
````

### Delete at Index

Traverse to the predecessor of the target node and bypass it.

**Time:** O(n). **Space:** O(1).

#### Steps

1. Delete node at index 1 in list `1 -> 3 -> 5`:
```text
before:  1 -> 3 -> 5 -> None
         ^    ^
         pred target

after:   1 -------> 5 -> None
         pred.next = target.next
```

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{removeAt} \\
\hline
\textbf{Input: } \text{int } index \\
\textbf{Output: } \text{node at position } index \text{ removed} \\
\\
\textbf{function } \text{removeAt}(index): \\
\quad 1: \quad \textbf{if } index \leq 0: \; head \leftarrow head.next \\
\quad 2: \quad \text{traverse to predecessor: } cur \leftarrow \text{node at position } index - 1 \\
\quad 3: \quad \text{bypass target: } cur.next \leftarrow cur.next.next \\
\hline
\end{array}
$$

````python
def removeAt(self, index):
    if index <= 0 and self.head:
        self.head = self.head.next
        return
    cnt = 0
    cur = self.head
    while cur and cur.next and cnt < index - 1:
        cur = cur.next
        cnt += 1
    if cur and cur.next:
        cur.next = cur.next.next
````

### Operations Summary

| Operation | Time | Space | Notes |
|-----------|------|-------|-------|
| Create | O(n) | O(n) | Build from sequence |
| Length | O(n) | O(1) | Full traversal |
| Search | O(n) | O(1) | Linear scan |
| Insert | O(n) | O(1) | O(1) once predecessor found |
| Update | O(n) | O(1) | Traverse to position |
| Delete | O(n) | O(1) | O(1) once predecessor found |

## 3. Linked Lists vs Arrays

| Property | Linked List | Array |
|----------|-------------|-------|
| Storage | Non-contiguous, pointer-linked | Contiguous block |
| Random access | O(n) traversal | O(1) by index |
| Insert/Delete (known pos) | O(1) pointer update | O(n) shift elements |
| Space overhead | Extra pointer per node | None beyond data |
| Memory allocation | Dynamic, per node | Fixed or amortized resize |

## 4. Two-Pointer Techniques

Two-pointer techniques on linked lists use two references moving through the list simultaneously. The movement strategy depends on the problem.

- **Fast/slow (offset start):** fast pointer leads by k steps, then both move together
- **Fast/slow (different speed):** slow moves 1 step, fast moves 2 steps per iteration
- **Separate pointers:** one pointer per list, used for merging or comparison

### Offset Start: Find k-th From End

The fast pointer advances k steps first. Then both move together until fast reaches the end. The slow pointer lands on the k-th node from the end.

**Time:** O(n). **Space:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{findKthFromEnd} \\
\hline
\textbf{Input: } \text{head node, int } k \\
\textbf{Output: } k\text{-th node from the tail} \\
\\
\textbf{function } \text{findKthFromEnd}(head, k): \\
\quad 1: \quad slow \leftarrow head, \; fast \leftarrow head \\
\quad 2: \quad \text{advance fast pointer } k \text{ nodes: } fast \leftarrow fast.next \text{ (repeat } k \text{ times)} \\
\quad 3: \quad \textbf{while } fast \neq \text{None}: \\
\quad 4: \quad\quad slow \leftarrow slow.next, \; fast \leftarrow fast.next \\
\quad 5: \quad \textbf{return } slow \\
\hline
\end{array}
$$

````python
def findKthFromEnd(head, k):
    slow = fast = head
    for _ in range(k):
        fast = fast.next
    while fast:
        slow = slow.next
        fast = fast.next
    return slow
````

#### Steps

```text
list: 1 -> 2 -> 3 -> 4 -> 5, k = 2

step 1: fast advances 2 steps
  slow         fast
   v             v
   1 -> 2 -> 3 -> 4 -> 5

step 2: both move together until fast = None
             slow         fast
               v             v
   1 -> 2 -> 3 -> 4 -> 5 -> None

slow points to node 4 (2nd from end)
```

### Different Speed: Find Middle Node

Slow moves 1 step, fast moves 2 steps. When fast reaches the end, slow is at the middle.

**Time:** O(n). **Space:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{findMiddle} \\
\hline
\textbf{Input: } \text{head node} \\
\textbf{Output: } \text{middle node of the list} \\
\\
\textbf{function } \text{findMiddle}(head): \\
\quad 1: \quad slow \leftarrow head, \; fast \leftarrow head \\
\quad 2: \quad \textbf{while } fast \neq \text{None} \textbf{ and } fast.next \neq \text{None}: \\
\quad 3: \quad\quad slow \leftarrow slow.next \\
\quad 4: \quad\quad fast \leftarrow fast.next.next \\
\quad 5: \quad \textbf{return } slow \text{ (middle position)} \\
\hline
\end{array}
$$

````python
def findMiddle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow
````

#### Steps

```text
list: 1 -> 2 -> 3 -> 4 -> 5

iteration 1: slow=2, fast=3
iteration 2: slow=3, fast=5
fast.next is None -> stop

slow points to node 3 (middle)
```

### Different Speed: Detect Cycle

If the list has a cycle, the fast pointer eventually catches the slow pointer. If fast reaches None, there is no cycle.

**Time:** O(n). **Space:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{hasCycle} \\
\hline
\textbf{Input: } \text{head node} \\
\textbf{Output: } \text{True if cycle exists, else False} \\
\\
\textbf{function } \text{hasCycle}(head): \\
\quad 1: \quad slow \leftarrow head, \; fast \leftarrow head \\
\quad 2: \quad \textbf{while } fast \neq \text{None} \textbf{ and } fast.next \neq \text{None}: \\
\quad 3: \quad\quad slow \leftarrow slow.next, \; fast \leftarrow fast.next.next \\
\quad 4: \quad\quad \textbf{if } slow = fast: \textbf{ return } \text{True} \\
\quad 5: \quad \textbf{return } \text{False (} fast \text{ reached tail)} \\
\hline
\end{array}
$$

````python
def hasCycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
````

### Separate Pointers: Merge Two Sorted Lists

One pointer per list. At each step, append the smaller node to the result and advance that pointer.

**Time:** O(n + m). **Space:** O(1).

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{mergeTwoLists} \\
\hline
\textbf{Input: } \text{sorted heads } l1, l2 \\
\textbf{Output: } \text{single sorted list} \\
\\
\textbf{function } \text{mergeTwoLists}(l1, l2): \\
\quad 1: \quad dummy \leftarrow \text{new node}, \; cur \leftarrow dummy \\
\quad 2: \quad \textbf{while } l1 \neq \text{None} \textbf{ and } l2 \neq \text{None}: \\
\quad 3: \quad\quad \textbf{if } l1.val \leq l2.val: \; cur.next \leftarrow l1, \; l1 \leftarrow l1.next \\
\quad 4: \quad\quad \textbf{else: } cur.next \leftarrow l2, \; l2 \leftarrow l2.next \\
\quad 5: \quad\quad cur \leftarrow cur.next \\
\quad 6: \quad cur.next \leftarrow \text{remaining nodes from } l1 \text{ or } l2 \\
\quad 7: \quad \textbf{return } dummy.next \\
\hline
\end{array}
$$

````python
def mergeTwoLists(l1, l2):
    dummy = ListNode(-1)
    cur = dummy
    while l1 and l2:
        if l1.val <= l2.val:
            cur.next = l1
            l1 = l1.next
        else:
            cur.next = l2
            l2 = l2.next
        cur = cur.next
    cur.next = l1 if l1 else l2
    return dummy.next
````

## 5. Sorting Linked Lists

Linked lists do not support random access, so not all sorting algorithms adapt well.

| Algorithm | Time | Space | Stable | Notes |
|-----------|------|-------|--------|-------|
| Bubble sort | O(n^2) | O(1) | Yes | Swap adjacent values |
| Selection sort | O(n^2) | O(1) | No | Find min, swap values |
| Insertion sort | O(n^2) | O(1) | Yes | Good for nearly sorted |
| Merge sort | O(n log n) | O(log n) | Yes | Best general choice |
| Quick sort | O(n log n) avg | O(log n) | No | Partition by pivot |
| Counting sort | O(n + k) | O(k) | Yes | Small value range |
| Bucket sort | O(n + k) | O(n + k) | Yes | Uniform distribution |
| Radix sort | O(n * d) | O(n + k) | Yes | Fixed digit count |

### Bubble Sort

Compare adjacent nodes and swap values if out of order. After each pass, the largest unsorted value bubbles to the tail.

**Time:** O(n^2). **Space:** O(1). **Stable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{bubbleSort} \\
\hline
\textbf{function } \text{bubbleSort}(head): \\
\quad 1: \quad tail \leftarrow \text{None (sorted boundary)} \\
\quad 2: \quad \textbf{while } head \neq tail: \\
\quad 3: \quad\quad cur \leftarrow head \\
\quad 4: \quad\quad \textbf{while } cur.next \neq tail: \\
\quad 5: \quad\quad\quad \textbf{if } cur.val > cur.next.val: \text{ swap values} \\
\quad 6: \quad\quad\quad cur \leftarrow cur.next \\
\quad 7: \quad\quad tail \leftarrow cur \text{ (largest in position)} \\
\quad 8: \quad \textbf{return } head \\
\hline
\end{array}
$$

````python
def bubbleSort(head):
    if not head or not head.next:
        return head
    tail = None
    while head != tail:
        cur = head
        while cur.next != tail:
            if cur.val > cur.next.val:
                cur.val, cur.next.val = cur.next.val, cur.val
            cur = cur.next
        tail = cur
    return head
````

### Selection Sort

Find the minimum value in the unsorted portion and swap it with the current front node.

**Time:** O(n^2). **Space:** O(1). **Unstable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{selectionSort} \\
\hline
\textbf{function } \text{selectionSort}(head): \\
\quad 1: \quad cur \leftarrow head \\
\quad 2: \quad \textbf{while } cur \neq \text{None}: \\
\quad 3: \quad\quad minNode \leftarrow cur \\
\quad 4: \quad\quad \textbf{for } \text{each node } scan \text{ from } cur.next \text{ to tail}: \\
\quad 5: \quad\quad\quad \textbf{if } scan.val < minNode.val: \; minNode \leftarrow scan \\
\quad 6: \quad\quad \textbf{if } minNode \neq cur: \text{ swap } cur.val, minNode.val \\
\quad 7: \quad\quad cur \leftarrow cur.next \\
\quad 8: \quad \textbf{return } head \\
\hline
\end{array}
$$

````python
def selectionSort(head):
    cur = head
    while cur and cur.next:
        minNode = cur
        scan = cur.next
        while scan:
            if scan.val < minNode.val:
                minNode = scan
            scan = scan.next
        if cur != minNode:
            cur.val, minNode.val = minNode.val, cur.val
        cur = cur.next
    return head
````

### Insertion Sort

Maintain a sorted prefix. For each unsorted node, detach it and insert it into the correct position in the sorted prefix.

**Time:** O(n^2). **Space:** O(1). **Stable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{insertionSort} \\
\hline
\textbf{function } \text{insertionSort}(head): \\
\quad 1: \quad dummy.next \leftarrow head, \; tail \leftarrow head \\
\quad 2: \quad cur \leftarrow head.next \\
\quad 3: \quad \textbf{while } cur \neq \text{None}: \\
\quad 4: \quad\quad \textbf{if } tail.val \leq cur.val: \; tail \leftarrow tail.next \\
\quad 5: \quad\quad \textbf{else:} \\
\quad 6: \quad\quad\quad \text{detach } cur: \; tail.next \leftarrow cur.next \\
\quad 7: \quad\quad\quad \text{find } prev \text{ in sorted portion where } prev.next.val > cur.val \\
\quad 8: \quad\quad\quad \text{insert: } cur.next \leftarrow prev.next, \; prev.next \leftarrow cur \\
\quad 9: \quad\quad cur \leftarrow tail.next \\
\quad 10: \quad \textbf{return } dummy.next \\
\hline
\end{array}
$$

````python
def insertionSort(head):
    if not head or not head.next:
        return head
    dummy = ListNode(-1)
    dummy.next = head
    tail = head
    cur = head.next
    while cur:
        if tail.val <= cur.val:
            tail = tail.next
        else:
            prev = dummy
            while prev.next.val <= cur.val:
                prev = prev.next
            tail.next = cur.next
            cur.next = prev.next
            prev.next = cur
        cur = tail.next
    return dummy.next
````

### Merge Sort

Split the list in half using fast/slow pointers, recursively sort each half, then merge.

**Time:** O(n log n). **Space:** O(log n) call stack. **Stable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{mergeSort} \\
\hline
\textbf{function } \text{mergeSort}(head): \\
\quad 1: \quad \textbf{if } head = \text{None or } head.next = \text{None}: \textbf{ return } head \\
\quad 2: \quad \text{find midpoint: } slow, fast \leftarrow head, head.next \\
\quad 3: \quad \text{split: } mid \leftarrow slow.next, \; slow.next \leftarrow \text{None} \\
\quad 4: \quad left \leftarrow \text{mergeSort}(head) \\
\quad 5: \quad right \leftarrow \text{mergeSort}(mid) \\
\quad 6: \quad \text{merge } left, right \text{ by comparing } .val \text{ ascending} \\
\quad 7: \quad \textbf{return } \text{merged head} \\
\hline
\end{array}
$$

````python
def mergeSort(head):
    if not head or not head.next:
        return head
    slow, fast = head, head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    mid = slow.next
    slow.next = None
    left = mergeSort(head)
    right = mergeSort(mid)
    dummy = ListNode(-1)
    cur = dummy
    while left and right:
        if left.val <= right.val:
            cur.next = left
            left = left.next
        else:
            cur.next = right
            right = right.next
        cur = cur.next
    cur.next = left if left else right
    return dummy.next
````

#### Steps

```text
list: 4 -> 2 -> 1 -> 3

split: [4 -> 2] and [1 -> 3]
split: [4] [2] and [1] [3]
merge: [2 -> 4] and [1 -> 3]
merge: [1 -> 2 -> 3 -> 4]
```

### Quick Sort

Choose a pivot (head value). Partition by scanning with two pointers: values less than pivot go left of the pivot position. Recurse on both sides.

**Time:** O(n log n) avg, O(n^2) worst. **Space:** O(log n). **Unstable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{quickSort} \\
\hline
\textbf{function } \text{partition}(left, right): \\
\quad 1: \quad pivot \leftarrow left.val \\
\quad 2: \quad slow \leftarrow left, \; fast \leftarrow left.next \\
\quad 3: \quad \textbf{while } fast \neq right: \\
\quad 4: \quad\quad \textbf{if } fast.val < pivot: \\
\quad 5: \quad\quad\quad slow \leftarrow slow.next \\
\quad 6: \quad\quad\quad \text{swap } slow.val, fast.val \\
\quad 7: \quad\quad fast \leftarrow fast.next \\
\quad 8: \quad \text{swap } slow.val, left.val \text{ (pivot to sorted position)} \\
\quad 9: \quad \textbf{return } slow \\
\\
\textbf{function } \text{quickSort}(left, right): \\
\quad 1: \quad pi \leftarrow \text{partition}(left, right) \\
\quad 2: \quad \text{quickSort}(left, pi) \\
\quad 3: \quad \text{quickSort}(pi.next, right) \\
\hline
\end{array}
$$

````python
def partition(left, right):
    if left == right or left.next == right:
        return left
    pivot = left.val
    slow, fast = left, left.next
    while fast != right:
        if fast.val < pivot:
            slow = slow.next
            slow.val, fast.val = fast.val, slow.val
        fast = fast.next
    slow.val, left.val = left.val, slow.val
    return slow


def quickSort(left, right):
    if left == right or left.next == right:
        return left
    pi = partition(left, right)
    quickSort(left, pi)
    quickSort(pi.next, right)
    return left
````

### Counting Sort

Count occurrences of each value, then rebuild the list in order. Requires a small, known value range.

**Time:** O(n + k). **Space:** O(k). **Stable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{countingSort} \\
\hline
\textbf{function } \text{countingSort}(head): \\
\quad 1: \quad \text{traverse list to find } lo, hi \\
\quad 2: \quad counts \leftarrow [0] \times (hi - lo + 1) \\
\quad 3: \quad \text{traverse list: } counts[cur.val - lo] \mathrel{+}= 1 \\
\quad 4: \quad \text{rebuild list from } counts \text{ in ascending order} \\
\quad 5: \quad \textbf{return } dummy.next \\
\hline
\end{array}
$$

````python
def countingSort(head):
    if not head or not head.next:
        return head
    lo, hi = float('inf'), float('-inf')
    cur = head
    while cur:
        lo = min(lo, cur.val)
        hi = max(hi, cur.val)
        cur = cur.next
    counts = [0] * (hi - lo + 1)
    cur = head
    while cur:
        counts[cur.val - lo] += 1
        cur = cur.next
    dummy = ListNode(-1)
    cur = dummy
    for i in range(len(counts)):
        while counts[i] > 0:
            cur.next = ListNode(i + lo)
            cur = cur.next
            counts[i] -= 1
    return dummy.next
````

### Bucket Sort

Distribute elements into buckets by value range. Sort each bucket (e.g. with merge sort), then concatenate.

**Time:** O(n) avg. **Space:** O(n + k). **Stable.**

### Radix Sort

Sort by each digit position from least significant to most significant, distributing into 10 buckets (0-9) per pass.

**Time:** O(n * d) where d = max digits. **Space:** O(n + k). **Stable.**

#### Pseudocode

$$
\begin{array}{l}
\hline
\textbf{Pattern: } \text{radixSort} \\
\hline
\textbf{function } \text{radixSort}(head): \\
\quad 1: \quad maxDigits \leftarrow \text{digits in largest value} \\
\quad 2: \quad \textbf{for } d \leftarrow 0 \text{ to } maxDigits - 1: \\
\quad 3: \quad\quad buckets \leftarrow 10 \text{ empty lists (digits 0..9)} \\
\quad 4: \quad\quad \text{traverse list: place each node in } buckets[(cur.val \;/\; 10^d) \bmod 10] \\
\quad 5: \quad\quad \text{concatenate } buckets[0] \text{ through } buckets[9] \text{ to rebuild list} \\
\quad 6: \quad \textbf{return } head \\
\hline
\end{array}
$$

````python
def radixSort(head):
    if not head:
        return head
    maxDigits = 0
    cur = head
    while cur:
        maxDigits = max(maxDigits, len(str(cur.val)))
        cur = cur.next
    for d in range(maxDigits):
        buckets = [[] for _ in range(10)]
        cur = head
        while cur:
            digit = (cur.val // (10 ** d)) % 10
            buckets[digit].append(cur.val)
            cur = cur.next
        dummy = ListNode(-1)
        cur = dummy
        for bucket in buckets:
            for val in bucket:
                cur.next = ListNode(val)
                cur = cur.next
        head = dummy.next
    return head
````

### Two-Pointer Patterns

| Pattern | Movement | Use-case |
