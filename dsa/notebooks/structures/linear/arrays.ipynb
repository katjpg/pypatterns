# Arrays & Sequences

## 1. Array Fundamentals

An array is a contiguous block of memory that stores elements sequentially and supports `O(1)` random access by index.

The address of `arr[i]` is `base + i * elementSize` where:
- `base` is the address of `arr[0]`. 
- `i` is the index.
- `elementSize` is the size of each element in bytes.

For example, if `base = 1000` and `elementSize = 4`, then `arr[0]` is at `1000`, `arr[1]` is at `1004`, and `arr[2]` is at `1008`. Each index increases the address by `elementSize` bytes.

Arrays can be defined by two types:
- A **static array** has fixed size after allocation.
- A **dynamic array** can grow by allocating a larger contiguous block and copying the existing elements.

Python’s `list` is a dynamic array of object references, so it supports `O(1)` index access and amortized `O(1)` append.

A sequence is an ordered collection that supports indexing, length, and iteration; common Python sequence types include `list`, `tuple`, `str`, and `range`.

A string (`str`) is an immutable character sequence, so many array patterns apply directly to strings.

## 2. Operations

Array problem solving is built on a small set of primitive operations: **non-mutating** operations read or copy the array, while **mutating** operations modify it in place.
Higher-level data structures are built from these same primitives:
- **Stacks** rely on fast `append` and `pop()` at the end.
- **Array-backed lists** rely on indexed access, indexed update, and resizing.
- **Sorting and searching algorithms** rely on traversal, comparison, swapping, slicing, and reordering.
- **Strings and grids** reuse the same sequence ideas: indexing, slicing, scanning, and range-based access.

### Performance Notes
- **Read operations** such as `len(nums)` and `nums[i]` run in constant `O(1)` time.
- **Slicing** creates a new list, so `nums[left:right]` costs `O(k)`.
- **Middle insertions and deletions** cost `O(n)` because later elements must shift.
- **Append** and `pop()` from the end are the main efficient mutating operations on Python lists.
- **Search operations** on arrays are linear unless another structure, such as sorting or hashing, is added.
- **Reverse** mutates in place, while `nums[::-1]` creates a new list.
- **Sort** is a preprocessing step that often simplifies array patterns.

### Non-mutating operations
- Read the array without changing it.
- May return a value or create a new sequence.
- Examples:
  - `len(nums)`
  - `nums[i]`
  - `nums[left:right]`
  - `val in nums`
  - `nums.index(val)`
  - `nums.count(val)`
  - `nums1 + nums2`
  - `nums * c`
  - `nums1 == nums2`

### Mutating operations
- Change the original array in place.
- Often used when the goal is `O(1)` extra space.
- Examples:
  - `nums[i] = val`
  - `nums.append(val)`
  - `nums.insert(i, val)`
  - `nums.extend(other)`
  - `nums.pop()`
  - `nums.pop(i)`
  - `del nums[i]`
  - `nums.remove(val)`
  - `nums.reverse()`
  - `nums.sort()`

## 3. Updates, Insertions, and Deletions

### Fast/Slow Two-Pointers

The **fast/slow pointer** (read/write pointer) technique uses two indices moving in the same direction:

- **`fast`** (read): scans every element, decides what to write.
- **`slow`** (write): tracks where the next valid element goes.

`fast` reads ahead, `slow` marks the write position. When `fast` finds a valid element, it writes to `slow` and both advance. Otherwise only `fast` advances. The result is always `nums[0:slow]`.

Deleting from the middle of an array costs `O(n)` per element (shift left). Removing `k` elements naively costs `O(nk)`. Fast/slow batches all removals into **one `O(n)` pass**, `fast` reads, `slow` writes only valid elements.

| Property | Detail |
|---|---|
| **Key idea** | Partition into "valid" and "unprocessed" regions with two same-direction pointers |
| **Use-case** | In-place deletion, deduplication, compaction, partitioning |
| **Problem keywords** | "remove", "in-place", "move", "compact", "deduplicate", "O(1) extra space" |
| **Time** | `O(n)` (single pass) |
| **Space** | `O(1)` (pointer variables) |

**Pointer motion variants:**

| Variant | Input Structure | Pointers | Use-case |
|---|---|---|---|
| Fast / slow | `nums = [3, 2, 2, 3], val = 3` | `slow, fast` both move left -> right | Deletion, compaction, dedup |
| Left / right (converging) | `nums = [1, 2, 3, 4, 5], left = 0, right = 4` | `left` -> <- `right` converge inward | Reversal, palindrome, sorted two-sum |
| Merge | `a = [1, 3, 5], b = [2, 4, 6]` | `i` advances through `a`, `j` through `b` | Merging sorted sequences |

```python
# Pattern - In-place Deletion
# time: O(n)
# space: O(1)


def removeElement(nums, target):
    slow = 0
    for fast in range(len(nums)):
        if nums[fast] != target:
            nums[slow] = nums[fast]
            slow += 1
    return slow
```

```python
# Pattern - Swap Partition
# time: O(n)
# space: O(1)


def moveZeroes(nums):
    slow = 0
    for fast in range(len(nums)):
        if nums[fast] != 0:
            nums[slow], nums[fast] = nums[fast], nums[slow]
            slow += 1
```

```python
# Pattern - In-place Deduplication
# time: O(n)
# space: O(1)


def removeDuplicates(nums):
    if not nums:
        return 0
    slow = 0
    for fast in range(1, len(nums)):
        if nums[fast] != nums[slow]:
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1
```

```python
# Pattern - Two-pointer Merge
# time: O(n + m)
# space: O(n + m)


def mergeSortedArrays(a, b):
    res = []
    i, j = 0, 0
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1
    res.extend(a[i:])
    res.extend(b[j:])
    return res
```

```python
# Pattern - Left/Right converging (in-place reversal)
# time: O(n)
# space: O(1)


def reverseArray(arr, left, right):
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
```

### Array Rotation

**Triple reversal** rotates an array in-place by reusing `reverseArray`:

- Reverse the entire array
- Reverse first `k` elements
- Reverse remaining `n - k` elements
- Left rotation by `k` = right rotation by `n - k`

```python
# Pattern - Triple reversal (in-place rotation)
# time: O(n)
# space: O(1)


def rotateArray(nums, k):
    n = len(nums)
    k %= n
    reverseArray(nums, 0, n - 1)
    reverseArray(nums, 0, k - 1)
    reverseArray(nums, k, n - 1)
```

## 4. Searching and Sorting

- **Searching** finds an element or position in an array. Linear scan is `O(n)`; binary search is `O(log n)` on sorted data.
- **Binary search variants** (`lowerBound`, `upperBound`, `firstTrue`) reduce different problems to the same halving template, only the condition changes.
- **Binary search on answer** applies `firstTrue` to optimization problems where the answer space is monotonic.
- **Sorting** reorders elements for faster lookup, deduplication, or greedy algorithms. Comparison sorts are `O(n log n)` at best; simple quadratic sorts are useful for small or nearly sorted inputs.

### Linear Search

- Scan every element left to right until `target` is found
- Return `i` when `arr[i] == target`, or `-1` if not found

| Property | Detail |
|---|---|
| **Input structure** | `i = 0` |
| **Key idea** | Sequential scan, check every element |
| **Use-case** | Search when array is unsorted or small |
| **Problem keywords** | "find", "contains", "exists", "check if present" |
| **Time** | `O(n)` |
| **Space** | `O(1)` |

```python
# Pattern - Linear Search
# time: O(n)
# space: O(1)


def linearSearch(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

### Binary Search

- Requires a **sorted** array
- Maintain `lo = 0`, `hi = n - 1`
- Compute `mid = lo + (hi - lo) // 2`
- `nums[mid] == target` -> found
- `nums[mid] < target` -> search right: `lo = mid + 1`
- `nums[mid] > target` -> search left: `hi = mid - 1`

| Property | Detail |
|---|---|
| **Input structure** | `lo = 0, hi = n - 1`, compute `mid` each iteration |
| **Key idea** | Compare `nums[mid]` to target; discard the half that cannot contain it |
| **Use-case** | Fast lookup in sorted data |
| **Problem keywords** | "sorted array", "find target", "search in sorted" |
| **Time** | `O(log n)` |
| **Space** | `O(1)` |
| **Loop invariant** | Target, if present, is in `nums[lo..hi]` |

```python
# Pattern - Binary Search
# time: O(log n)
# space: O(1)


def binarySearch(nums, target):
    lo, hi = 0, len(nums) - 1
    while lo <= hi:
        mid = lo + (hi - lo) // 2
        if nums[mid] == target:
            return mid
        if nums[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1
```

### Lower Bound

- First index where `nums[i] >= target` (equivalent to `bisect_left`)
- `hi = n` (not `n - 1`) answer can be past the last element
- `nums[mid] < target` -> `lo = mid + 1`
- Otherwise -> `hi = mid`

| Property | Detail |
|---|---|
| **Input structure** | `lo = 0, hi = n` (note `hi = n`, not `n - 1`) |
| **Key idea** | Binary search where `nums[mid] < target` narrows left |
| **Use-case** | Find insertion point, first occurrence of target |
| **Problem keywords** | "first position >=", "leftmost occurrence", "insert position" |
| **Time** | `O(log n)` |
| **Space** | `O(1)` |
| **Returns** | First `i` where `nums[i] >= target`, or `n` if all smaller |

```python
# Pattern - Lower Bound (bisect_left)
# time: O(log n)
# space: O(1)


def lowerBound(nums, target):
    lo, hi = 0, len(nums)
    while lo < hi:
        mid = lo + (hi - lo) // 2
        if nums[mid] < target:
            lo = mid + 1
        else:
            hi = mid
    return lo
```

### Upper Bound

- First index where `nums[i] > target` (equivalent to `bisect_right`)
- Same structure as lower bound, condition flips from `<` to `<=`
- `nums[mid] <= target` -> `lo = mid + 1`
- Otherwise -> `hi = mid`

| Property | Detail |
|---|---|
| **Input structure** | `lo = 0, hi = n` (same bounds as lower bound) |
| **Key idea** | Binary search where `nums[mid] <= target` narrows left |
| **Use-case** | Find end of target range, count occurrences |
| **Problem keywords** | "first position >", "rightmost occurrence", "count of target" |
| **Time** | `O(log n)` |
| **Space** | `O(1)` |
| **Returns** | First `i` where `nums[i] > target`, or `n` if all `<= target` |

**Lower bound vs. upper bound:**

| Function | Condition | Finds |
|---|---|---|
| `lowerBound` | `nums[mid] < target` | First `i` where `nums[i] >= target` |
| `upperBound` | `nums[mid] <= target` | First `i` where `nums[i] > target` |

Count of `target` in sorted array = `upperBound - lowerBound`.

```python
# Pattern - Upper Bound (bisect_right)
# time: O(log n)
# space: O(1)


def upperBound(nums, target):
    lo, hi = 0, len(nums)
    while lo < hi:
        mid = lo + (hi - lo) // 2
        if nums[mid] <= target:
            lo = mid + 1
        else:
            hi = mid
    return lo
```

### First True (Binary Search on Condition)

- Generalize binary search to any monotonic boolean predicate `f`
- Search space: `[False, False, ..., True, True, ...]`
- `f(mid)` is `True` -> answer could be `mid` or earlier: `hi = mid`
- `f(mid)` is `False` -> answer is right: `lo = mid + 1`

`lowerBound` and `upperBound` are special cases:
- `lowerBound(nums, t)` = `firstTrue(0, n, lambda i: nums[i] >= t)`
- `upperBound(nums, t)` = `firstTrue(0, n, lambda i: nums[i] > t)`

**Binary search on answer** applies `firstTrue` to optimization problems. `splitArray(nums, k)` minimizes the maximum subarray sum when splitting into `k` parts. Predicate: "can we split into `<= k` parts with max sum `<= mid`?" Search space: `[max(nums), sum(nums)]`.

```python
# Pattern - First True (binary search on condition)
# time: O(log(hi - lo) * cost of f)
# space: O(1)


def firstTrue(lo, hi, f):
    while lo < hi:
        mid = lo + (hi - lo) // 2
        if f(mid):
            hi = mid
        else:
            lo = mid + 1
    return lo
```

```python
# Pattern - Binary Search on Answer
# time: O(n * log(sum - max))
# space: O(1)


def splitArray(nums, k):
    def canSplit(maxSum):
        parts, total = 1, 0
        for val in nums:
            if total + val > maxSum:
                parts += 1
                total = val
            else:
                total += val
        return parts <= k

    return firstTrue(max(nums), sum(nums), canSplit)
```

### Insertion Sort

- Build a sorted prefix one element at a time
- Take `key = arr[i]`, shift elements right while `arr[j] > key`
- Place `key` in the gap

| Property | Detail |
|---|---|
| **Input structure** | `key = arr[i], j = i - 1`, scan backward to find insertion point |
| **Key idea** | Grow a sorted prefix by inserting each element into its correct position |
| **Use-case** | Small arrays, nearly sorted data, online sorting |
| **Problem keywords** | "sort", "nearly sorted", "k elements away", "online insertion" |
| **Time** | `O(n)` best (sorted), `O(n^2)` worst (reverse sorted) |
| **Space** | `O(1)` |

```python
# Pattern - Insertion Sort
# time: O(n^2)
# space: O(1)


def insertionSort(arr):
    res = arr[:]
    for i in range(1, len(res)):
        key = res[i]
        j = i - 1
        while j >= 0 and res[j] > key:
            res[j + 1] = res[j]
            j -= 1
        res[j + 1] = key
    return res
```

### Selection Sort

- Find the minimum in the unsorted region, swap to front
- Sorted prefix grows by one each pass

| Property | Detail |
|---|---|
| **Input structure** | `minIdx = i`, scan forward to find the minimum |
| **Key idea** | Repeatedly select the minimum from the unsorted region |
| **Use-case** | When minimizing swaps matters, small arrays |
| **Problem keywords** | "sort", "minimum swaps", "selection" |
| **Time** | `O(n^2)` (always) |
| **Space** | `O(1)` |

```python
# Pattern - Selection Sort
# time: O(n^2)
# space: O(1)


def selectionSort(arr):
    res = arr[:]
    n = len(res)
    for i in range(n - 1):
        minIdx = i
        for j in range(i + 1, n):
            if res[j] < res[minIdx]:
                minIdx = j
        res[i], res[minIdx] = res[minIdx], res[i]
    return res
```

### Merge Sort

- Divide array in half, recursively sort each half, merge
- Base case: length `<= 1`
- Reuses `mergeSortedArrays` from Section 3

| Property | Detail |
|---|---|
| **Input structure** | `mid = len(arr) // 2`, split into `arr[:mid]` and `arr[mid:]` |
| **Key idea** | Divide in half, sort each half, merge the two sorted halves |
| **Use-case** | `O(n log n)` sort, suitable for linked lists |
| **Problem keywords** | "sort", "merge two sorted", "count inversions" |
| **Time** | `O(n log n)` (always) |
| **Space** | `O(n)` |

```python
# Pattern - Merge Sort (divide and conquer)
# time: O(n log n)
# space: O(n)


def mergeSort(arr):
    if len(arr) <= 1:
        return arr[:]
    mid = len(arr) // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    return mergeSortedArrays(left, right)
```

## 5. Precomputation Techniques

Precomputation is a one-pass scan over the array that stores useful values (frequency counts, prefix/suffix min and max, or cumulative sums) so each later query is an `O(1)` lookup instead of a repeated scan.

- **Frequency arrays** and **hash maps** count how often each value appears.
- **Prefix/suffix arrays** record the min or max seen so far from each end.
- **Prefix sums** record cumulative totals so any range sum is a single subtraction.

### Frequency Arrays

- A fixed-size array where `freq[c]` counts occurrences of character or value `c`
- Works when the value range is small and known (e.g., 26 lowercase letters, digits 0–9)
- Single pass to build, `O(1)` lookup per query

| Property | Detail |
|---|---|
| **Input structure** | `freq = [0] * 26`, fixed-size array of zeros |
| **Key idea** | Map bounded values to array indices, increment counts |
| **Use-case** | Character counting, anagram detection, histogram |
| **Problem keywords** | "count characters", "anagram", "permutation", "histogram" |
| **Time** | `O(n)` to build |
| **Space** | `O(k)` where `k` is the value range |

```python
# Pattern - Frequency Array (character counting)
# time: O(n)
# space: O(1)


def countChars(s):
    freq = [0] * 26
    for ch in s:
        freq[ord(ch) - ord("a")] += 1
    return freq


# Pattern - Anagram Check (frequency comparison)
# time: O(n)
# space: O(1)


def isAnagram(s, t):
    if len(s) != len(t):
        return False
    freq = [0] * 26
    for i in range(len(s)):
        freq[ord(s[i]) - ord("a")] += 1
        freq[ord(t[i]) - ord("a")] -= 1
    return all(c == 0 for c in freq)
```

### Hash-Map Counting

- Build a dictionary mapping each value to its count
- Any hashable key type works, not limited to small integer ranges
- Lookup, insert, and update are each `O(1)` amortized

| Property | Detail |
|---|---|
| **Input structure** | `freq = {}`, empty dictionary |
| **Key idea** | Dict-based frequency map for arbitrary values |
| **Use-case** | First unique element, top-k frequent, group by count |
| **Problem keywords** | "count occurrences", "first unique", "top k frequent", "most common" |
| **Time** | `O(n)` to build |
| **Space** | `O(n)` worst case |

```python
# Pattern - Hash-Map Frequency Build
# time: O(n)
# space: O(n)


def buildFreq(arr):
    freq = {}
    for val in arr:
        freq[val] = freq.get(val, 0) + 1
    return freq


# Pattern - First Unique Element
# time: O(n)
# space: O(n)


def firstUnique(arr):
    freq = buildFreq(arr)
    for val in arr:
        if freq[val] == 1:
            return val
    return -1


# Pattern - Top K Frequent (bucket sort)
# time: O(n)
# space: O(n)


def topKFrequent(nums, k):
    freq = buildFreq(nums)
    n = len(nums)
    buckets = [[] for _ in range(n + 1)]
    for val, cnt in freq.items():
        buckets[cnt].append(val)
    res = []
    for i in range(n, 0, -1):
        for val in buckets[i]:
            res.append(val)
            if len(res) == k:
                return res
    return res
```

### Precomputed Minima / Maxima

- Scan left to right: `prefixMax[i] = max(prefixMax[i-1], nums[i])`
- Scan right to left: `suffixMax[i] = max(suffixMax[i+1], nums[i])`
- The max or min at any index is then a direct `O(1)` read

| Property | Detail |
|---|---|
| **Input structure** | `res = [0] * n, res[0] = nums[0]` |
| **Key idea** | Store running min/max from each direction |
| **Use-case** | Position-based max/min queries after one pass |
| **Problem keywords** | "trapping rain water", "container with most water", "max seen so far" |
| **Time** | `O(n)` to build |
| **Space** | `O(n)` |

```python
# Pattern - Prefix Max
# time: O(n)
# space: O(n)


def prefixMax(nums):
    n = len(nums)
    res = [0] * n
    res[0] = nums[0]
    for i in range(1, n):
        res[i] = max(res[i - 1], nums[i])
    return res


# Pattern - Suffix Max
# time: O(n)
# space: O(n)


def suffixMax(nums):
    n = len(nums)
    res = [0] * n
    res[n - 1] = nums[n - 1]
    for i in range(n - 2, -1, -1):
        res[i] = max(res[i + 1], nums[i])
    return res
```

## 6. Prefix Sums

`prefix[i]` stores the sum of `nums[0..i-1]`, with `prefix[0] = 0`. The sum of `nums[left..right]` = `prefix[right + 1] - prefix[left]`.

`rangeSum(prefix, 1, 3)` = `prefix[4] - prefix[1]` = `10 - 1` = `9`

- **Difference arrays** invert prefix sums: each range update is two point operations, then one prefix-sum pass produces the result.
- **Hash-map prefix sums** store `{prefixSum: count}` to find subarrays summing to a target in `O(n)`.

```python
# Pattern - Build Prefix Sum
# time: O(n)
# space: O(n)


def buildPrefix(nums):
    n = len(nums)
    prefix = [0] * (n + 1)
    for i in range(n):
        prefix[i + 1] = prefix[i] + nums[i]
    return prefix


# Pattern - Range Sum Query
# time: O(1)
# space: O(1)


def rangeSum(prefix, left, right):
    return prefix[right + 1] - prefix[left]
```

### Difference Arrays

- Add `val` to `nums[left..right]`: set `diff[left] += val`, `diff[right + 1] -= val`
- Apply all range updates, then prefix-sum `diff` to get the final array
- `m` range updates on `n` elements: `O(n + m)` total

| Property | Detail |
|---|---|
| **Input structure** | `diff = [0] * (n + 1)` |
| **Key idea** | Store boundary deltas instead of updating every element |
| **Use-case** | Batch range updates in one pass |
| **Problem keywords** | "range add", "increment subarray", "batch update", "flight bookings" |
| **Time** | `O(n + m)` for `m` updates on size `n` |
| **Space** | `O(n)` |

```python
# Pattern - Difference Array (batch range updates)
# time: O(n + m)
# space: O(n)


def rangeUpdates(n, ops):
    diff = [0] * (n + 1)
    for left, right, val in ops:
        diff[left] += val
        diff[right + 1] -= val
    res = [0] * n
    res[0] = diff[0]
    for i in range(1, n):
        res[i] = res[i - 1] + diff[i]
    return res
```

### Prefix Sums on Binary Arrays and Strings

- Prefix sum on `[0, 1, 1, 0, 1]` counts 1s in any range: `prefix[right + 1] - prefix[left]`
- Replace 0 with -1 to convert "equal 0s and 1s" into "subarray sum = 0"
- Longest such subarray: track earliest index for each prefix sum in a hash map

| Property | Detail |
|---|---|
| **Input structure** | `hashmap = {0: -1}, prefixSum = 0` |
| **Key idea** | Prefix sum on binary data; 0 -> -1 transform for balance problems |
| **Use-case** | Count 1s in range, longest balanced subarray |
| **Problem keywords** | "equal 0s and 1s", "count 1s in range", "longest balanced subarray" |
| **Time** | `O(n)` build, `O(1)` query |
| **Space** | `O(n)` |

```python
# Pattern - Longest Subarray with Equal 0s and 1s
# time: O(n)
# space: O(n)


def findMaxLengthEqual(nums):
    hashmap = {0: -1}
    maxLen, prefixSum = 0, 0
    for i in range(len(nums)):
        prefixSum += 1 if nums[i] == 1 else -1
        if prefixSum in hashmap:
            maxLen = max(maxLen, i - hashmap[prefixSum])
        else:
            hashmap[prefixSum] = i
    return maxLen


# Pattern - Count 1s in Range (prefix sum on binary array)
# time: O(n) build, O(1) query
# space: O(n)


def countRangeOnes(nums, left, right):
    prefix = buildPrefix(nums)
    return rangeSum(prefix, left, right)
```

### Prefix Sums with Hash Maps (Subarray Sum Patterns)

- Maintain a running prefix sum and a hash map of previously seen sums
- If `currentPrefix - k` exists in the map, a subarray summing to `k` ends here
- `{sum: count}` for counting subarrays; `{sum: firstIndex}` for finding the longest

| Property | Detail |
|---|---|
| **Input structure** | `hashmap = {0: 1}, prefixSum = 0, cnt = 0` |
| **Key idea** | Hash map of prefix sums for O(1) complement lookups |
| **Use-case** | Count or find subarrays matching a sum condition |
| **Problem keywords** | "subarray sum equals k", "divisible by k", "pivot index", "contiguous subarray" |
| **Time** | `O(n)` |
| **Space** | `O(n)` |

```python
# Pattern - Count Subarrays Summing to K
# time: O(n)
# space: O(n)


def subarraySumK(nums, k):
    hashmap = {0: 1}
    prefixSum, cnt = 0, 0
    for val in nums:
        prefixSum += val
        cnt += hashmap.get(prefixSum - k, 0)
        hashmap[prefixSum] = hashmap.get(prefixSum, 0) + 1
    return cnt


# Pattern - Longest Subarray Summing to K
# time: O(n)
# space: O(n)


def longestSubarraySumK(nums, k):
    hashmap = {0: -1}
    prefixSum, maxLen = 0, 0
    for i in range(len(nums)):
        prefixSum += nums[i]
        if prefixSum - k in hashmap:
            maxLen = max(maxLen, i - hashmap[prefixSum - k])
        if prefixSum not in hashmap:
            hashmap[prefixSum] = i
    return maxLen


# Pattern - Count Subarrays Divisible by K
# time: O(n)
# space: O(n)


def subarrayDivByK(nums, k):
    hashmap = {0: 1}
    prefixSum, cnt = 0, 0
    for val in nums:
        prefixSum = (prefixSum + val) % k
        cnt += hashmap.get(prefixSum, 0)
        hashmap[prefixSum] = hashmap.get(prefixSum, 0) + 1
    return cnt


# Pattern - Pivot Index
# time: O(n)
# space: O(1)


def pivotIndex(nums):
    total = sum(nums)
    leftSum = 0
    for i in range(len(nums)):
        if leftSum == total - leftSum - nums[i]:
            return i
        leftSum += nums[i]
    return -1
```

## 7. Multidimensional Data Sets

A 2D array (matrix) is a grid of `rows x cols` elements accessed by two indices: `grid[row][col]`. Here, a matrix is a list of lists.

- **Construction** allocates a `rows x cols` grid filled with a default value.
- **Traversal patterns** visit every cell in a specific order: row-major, diagonal, or spiral.
- **2D prefix sums** extend 1D prefix sums to rectangular subregions.

### 2D Arrays and Matrix Representation

- `grid[row][col]` accesses the element at row `row`, column `col`
- A `rows x cols` grid has `rows * cols` elements
- Sorted matrix search starts from the top-right corner: go down if too small, go left if too large

`searchMatrix(grid, 9)` trace on a 4x4 sorted grid:

| Step | Position | Value | Action |
|---|---|---|---|
| 1 | `grid[0][3]` | 11 | 11 > 9, col-- |
| 2 | `grid[0][2]` | 7 | 7 < 9, row++ |
| 3 | `grid[1][2]` | 8 | 8 < 9, row++ |
| 4 | `grid[2][2]` | 9 | 9 == 9, found |

| Property | Detail |
|---|---|
| **Input structure** | `row = 0, col = cols - 1` |
| **Key idea** | Start top-right; go down if too small, left if too large |
| **Use-case** | Search in row-sorted and column-sorted matrix |
| **Problem keywords** | "search 2D matrix", "sorted rows and columns", "matrix search" |
| **Time** | `O(rows + cols)` |
| **Space** | `O(1)` |

```python
# Pattern - Grid Construction
# time: O(rows * cols)
# space: O(rows * cols)


def makeGrid(rows, cols, default=0):
    return [[default] * cols for _ in range(rows)]


# Pattern - Sorted Matrix Search (staircase)
# time: O(rows + cols)
# space: O(1)


def searchMatrix(grid, target):
    if not grid or not grid[0]:
        return False
    rows, cols = len(grid), len(grid[0])
    row, col = 0, cols - 1
    while row < rows and col >= 0:
        if grid[row][col] == target:
            return True
        if grid[row][col] < target:
            row += 1
        else:
            col -= 1
    return False
```

### Traversal Patterns

- **Row-major**: visit every row left to right, top to bottom
- **Diagonal**: group cells by `row + col`; each diagonal shares the same sum
- **Spiral**: peel layers from outside in, shrinking `top`, `bottom`, `left`, `right` boundaries each pass

| Pattern | Order | Time |
|---|---|---|
| Row-major | `(0,0), (0,1), ..., (1,0), (1,1), ...` | `O(rows * cols)` |
| Diagonal | `(0,0), (1,0), (0,1), (2,0), (1,1), (0,2), ...` | `O(rows * cols)` |
| Spiral | outer ring, then next inner ring, ... | `O(rows * cols)` |

```python
# Pattern - Row-Major Traversal
# time: O(rows * cols)
# space: O(rows * cols)


def traverse(grid):
    res = []
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            res.append(grid[row][col])
    return res


# Pattern - Diagonal Traversal
# time: O(rows * cols)
# space: O(rows * cols)


def diagonalGrid(grid):
    if not grid:
        return []
    rows, cols = len(grid), len(grid[0])
    res = []
    for d in range(rows + cols - 1):
        for row in range(max(0, d - cols + 1), min(d + 1, rows)):
            col = d - row
            res.append(grid[row][col])
    return res


# Pattern - Spiral Traversal
# time: O(rows * cols)
# space: O(rows * cols)


def spiralTraversal(grid):
    if not grid:
        return []
    res = []
    top, bottom, left, right = 0, len(grid) - 1, 0, len(grid[0]) - 1
    while top <= bottom and left <= right:
        for col in range(left, right + 1):
            res.append(grid[top][col])
        top += 1
        for row in range(top, bottom + 1):
            res.append(grid[row][right])
        right -= 1
        if top <= bottom:
            for col in range(right, left - 1, -1):
                res.append(grid[bottom][col])
            bottom -= 1
        if left <= right:
            for row in range(bottom, top - 1, -1):
                res.append(grid[row][left])
            left += 1
    return res
```

### 2D Prefix Sums

- `rowPrefix[r][c]` = sum of `grid[r][0..c-1]`, built independently per row
- Range sum within a single row: `rowPrefix[r][c2+1] - rowPrefix[r][c1]`
- `longestPrefix(strs)` finds the longest common prefix shared by all strings by shortening the candidate until every string matches

| Property | Detail |
|---|---|
| **Input structure** | `prefix = [[0] * (cols + 1) for _ in range(rows)]` |
| **Key idea** | Build prefix sums per row for O(1) row-range queries |
| **Use-case** | Row range sums, longest common prefix of strings |
| **Problem keywords** | "row sum", "submatrix sum", "longest common prefix" |
| **Time** | `O(rows * cols)` build, `O(1)` row query |
| **Space** | `O(rows * cols)` |

```python
# Pattern - Row Prefix Sums
# time: O(rows * cols)
# space: O(rows * cols)


def rowPrefixSums(grid):
    rows, cols = len(grid), len(grid[0])
    prefix = [[0] * (cols + 1) for _ in range(rows)]
    for r in range(rows):
        for c in range(cols):
            prefix[r][c + 1] = prefix[r][c] + grid[r][c]
    return prefix


# Pattern - Longest Common Prefix
# time: O(n * m)
# space: O(m)


def longestPrefix(strs):
    if not strs:
        return ""
    prefix = strs[0]
    for s in strs[1:]:
        while not s.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix
```
